## (s1p1)

```python
class gauss_s1p1(ThreeDScene):
    def construct(self):
        self.set_camera_orientation(phi=75 * DEGREES, theta=-45 * DEGREES)
        center = np.array([0, 0, 0])
        sphere_radius = 0.3
        sphere = Sphere(radius=sphere_radius, color=PURE_RED, stroke_width=0)
        sphere.set_fill(PURE_RED, opacity=1)
        self.add(sphere)

        # Generate directions using spherical coordinates
        num_arrows = 18
        arrow_length = 3
        arrows = VGroup()
        for i in range(num_arrows):
            theta = np.arccos(1 - 2 * (i + 0.5) / num_arrows)  # polar angle
            phi = np.pi * (1 + 5**0.5) * i  # golden angle for even spacing
            direction = np.array([
                np.sin(theta) * np.cos(phi),
                np.sin(theta) * np.sin(phi),
                np.cos(theta)
            ])
            start = center + sphere_radius * direction
            end = start + arrow_length * direction
            arrow = Arrow3D(
                start=start,
                end=end,
                color=YELLOW
            )
            arrows.add(arrow)
        self.add(arrows)
        self.wait(2)

        self.play(FadeOut(arrows), FadeOut(sphere))
        self.wait(1)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING gauss_s1p1
```

## (s1p3)

```python
class gauss_s1p3(Scene):
    def construct(self):
        num_arrows = 12
        circle_radius = 2
        arrow_length = 7
        dot_radius_min = 0.06
        dot_radius_max = 0.18
        max_dist = arrow_length
        coincidence_tol = 0.03  # threshold for "same point"

        circle = Circle(radius=circle_radius, color=BLUE, fill_opacity=0.4)
        charge_pos = np.array([0, 0, 0])
        charge = Dot(point=charge_pos, color=RED, radius=0.15)

        angles = [i * TAU / num_arrows for i in range(num_arrows)]
        directions = [np.array([np.cos(a), np.sin(a), 0]) for a in angles]

        arrows = VGroup()
        for direction in directions:
            arrow = always_redraw(
                lambda d=direction: Arrow(
                    start=charge.get_center(),
                    end=charge.get_center() + arrow_length * d,
                    color=YELLOW,
                    buff=0
                )
            )
            arrows.add(arrow)

        charge_group = VGroup(arrows, charge)

        def get_circle_intersections(charge_center, direction, circle_center, radius):
            d = direction / np.linalg.norm(direction)
            f = charge_center - circle_center
            a = np.dot(d, d)
            b = 2 * np.dot(f, d)
            c = np.dot(f, f) - radius**2
            discriminant = b**2 - 4*a*c
            if discriminant < 0:
                return []
            sqrt_disc = np.sqrt(discriminant)
            t1 = (-b + sqrt_disc) / (2*a)
            t2 = (-b - sqrt_disc) / (2*a)
            return [charge_center + t * d for t in [t1, t2] if t > 0]

        intersection_dots = VGroup()
        for direction in directions:
            # Use a closure to share state between the two updaters
            shared = {"positions": [None, None], "opacities": [1, 1]}

            def make_dot(idx, color, d=direction, shared=shared):
                dot = Dot(color=color, radius=0.2)
                def updater(dot):
                    intersections = get_circle_intersections(
                        charge.get_center(),
                        d,
                        circle.get_center(),
                        circle_radius
                    )
                    if len(intersections) > idx:
                        pos = intersections[idx]
                        dot.move_to(pos)
                        dist = np.linalg.norm(pos - charge.get_center())
                        r = dot_radius_max - (dot_radius_max - dot_radius_min) * min(dist / max_dist, 1)
                        dot.set_radius(r)
                        dot.set_opacity(1)
                        shared["positions"][idx] = pos
                    else:
                        dot.set_opacity(0)
                        shared["positions"][idx] = None

                    # Coincidence check: if both exist and are close, hide both
                    p0, p1 = shared["positions"]
                    if p0 is not None and p1 is not None and np.linalg.norm(p0 - p1) < coincidence_tol:
                        dot.set_opacity(0)
                dot.add_updater(updater)
                return dot

            intersection_dots.add(make_dot(0, PURE_GREEN))
            intersection_dots.add(make_dot(1, PURE_RED))

        line_path = Line(start=ORIGIN, end=RIGHT * 1.3)
        circ_path = Circle(radius=1.3)    

        self.add(circle, charge_group, intersection_dots)
        self.play(MoveAlongPath(charge_group, line_path, run_time=3))
        self.wait(0.5)
        self.play(MoveAlongPath(charge_group, circ_path, run_time=6))
        self.play(MoveAlongPath(charge_group, circ_path, run_time=6))
        self.wait(1)

        line_path_2 = Line(start=RIGHT*1.3, end=LEFT*1.3)
        line_path_2_reverse = Line(start=LEFT*1.3, end=RIGHT*1.3)
        self.play(MoveAlongPath(charge_group, line_path_2, run_time=3))
        self.play(MoveAlongPath(charge_group, line_path_2_reverse, run_time=3))
        self.wait(2)

        line_to_center = Line(start=RIGHT*1.3, end=ORIGIN)
        self.play(MoveAlongPath(charge_group, line_to_center, run_time=3))
        self.wait(2)




# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING gauss_s1p3
```

```python
class gauss_s2p1(Scene):
    def construct(self):

        charge = Dot(point=ORIGIN, color=RED, radius=0.15)
        circ = Circle(radius=2, color=BLUE, fill_opacity=0.4)



        group = VGroup(charge, circ)
        group.shift(LEFT * 3)

        self.play(FadeIn(charge))
        self.play(DrawBorderThenFill(circ))
        self.wait(1)

        flux_eqn_1 = MathTex(r"\Phi =", r"\oint", r"\vec{E}", r" \cdot ", r"d \vec{A}", r" \ \propto Q_{enc} ", color=WHITE).shift(RIGHT * 2.5)
        flux_eqn_2 = MathTex(r"\Phi =", r"\oint", r"\vec{E}", r" \cdot ", r"d \vec{A}", r" = \frac{Q_{enc}}{\varepsilon_0}", color=WHITE).shift(RIGHT * 2.5)


        self.play(Write(flux_eqn_1[0:5]))
        self.wait(2)

    
        self.play(FadeOut(flux_eqn_1[0:1]), Write(flux_eqn_1[5:6]))
        self.wait(2)

        num_dots = 25
        circle_center = circ.get_center()
        circle_radius = circ.radius*0.8

        dots = VGroup()
        for _ in range(num_dots):
            # Random radius (uniform area distribution)
            r = circle_radius * (random.random() ** 0.5)
            theta = random.uniform(0, TAU)
            x = circle_center[0] + r * np.cos(theta)
            y = circle_center[1] + r * np.sin(theta)
            color = random.choice([RED, BLUE])
            dot = Dot(point=[x, y, 0], color=color, radius=0.12)
            dots.add(dot)


        self.play(
            LaggedStart(
                *[FadeIn(dot) for dot in dots],
                lag_ratio=0.3
            )
        

        self.wait(1)

        dots_out = VGroup()
        for _ in range(num_dots):
            # Random radius (uniform area distribution)
            r = circle_radius*1.2 + (random.random() ** 0.5)
            theta = random.uniform(0, TAU)
            x = circle_center[0] + r * np.cos(theta)
            y = circle_center[1] + r * np.sin(theta)
            color = random.choice([RED, BLUE])
            dot = Dot(point=[x, y, 0], color=color, radius=0.12)
            dots_out.add(dot)

        self.play(
            LaggedStart(
                *[FadeIn(dot) for dot in dots_out],
                lag_ratio=0.3
            )
        )

        self.wait(2)
        self.play(
            LaggedStart(
                *[FadeOut(dot) for dot in dots_out],
                lag_ratio=0.2
            )
        )

        self.play(Transform(flux_eqn_1[5], flux_eqn_2[5]))

        self.wait(2)

        box = SurroundingRectangle(flux_eqn_1, color=YELLOW, buff=0.3).shift(RIGHT * 0.5)
        gauss_txt = Text("Gauss's Law", color=YELLOW).scale(1).next_to(box, UP*1.5)
        self.play(Create(box))
        self.wait(2)
        self.play(Write(gauss_txt))
        self.wait(2)

        self.play(Indicate(flux_eqn_1[1:5], color=PURE_GREEN, scale_factor=1.2, run_time=2))
        self.wait(2)
        self.play(Indicate(flux_eqn_1[5], color=PURE_GREEN, scale_factor=1.2, run_time=2))
        self.wait(2)

        circ2 = Circle(radius=3, color=BLUE, fill_opacity=0.4).shift(LEFT * 3)
        square = Square(side_length=3.5, color=BLUE, fill_opacity=0.4).shift(LEFT * 3)
        star = Star(color=BLUE, fill_opacity=0.4).shift(LEFT * 3).scale(4).rotate(-PI/4)

        self.play(ReplacementTransform(circ, circ2))
        self.wait(2)

        self.play(ReplacementTransform(circ2, circ))
        self.wait(2)

        self.play(ReplacementTransform(circ, square))
        self.wait(2)

        self.play(ReplacementTransform(square, star))
        self.wait(2)

        circ3 = Circle(radius=2, color=BLUE, fill_opacity=0.4).shift(LEFT * 3)
        self.play(ReplacementTransform(star, circ3))
        self.wait(2)

        self.play(Indicate(flux_eqn_1[2:3], color=PURE_GREEN, scale_factor=1.5, run_time=3))
        self.wait(2)


# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING gauss_s2p1
```
