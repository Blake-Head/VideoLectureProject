# Basic Streamline function

```python
class StreamLinesExample(Scene):
    def construct(self):
        # Define the vector field function
        # This function takes a position (pos) as a numpy array [x, y, z]
        # and returns a vector [dx, dy, dz] at that point.
        def vector_field_func(pos):
            # A simple rotational field: F(x, y) = -y*i + x*j
            return np.array([-pos[1], pos[0], 0])

        # Create the StreamLines mobject
        # It takes the vector field function as the first argument.
        stream_lines = StreamLines(
            vector_field_func,
            x_range=[-3, 3, 0.5],  # x-range with a step size for starting points
            y_range=[-3, 3, 0.5],  # y-range with a step size for starting points
            stroke_width=2,
            virtual_time=1.5,      # How long the "particles" flow
            max_anchors_per_line=30, # Smoother curves
        )

        # Add the streamlines to the scene and animate their creation
        self.play(Create(stream_lines))
        self.wait()

# load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING StreamLinesExample
```

# Electric Field Streamline Attempt

```python
class ElectricFieldAroundCircle(Scene):
    def construct(self):
        # --- Define constants and objects ---
        RADIUS = 1.0
        
        # The conducting circle in the center
        circle = Circle(
            radius=RADIUS,
            color=DARK_GRAY,
            fill_opacity=1.0
        )
        
        # --- Define the vector field function ---
        def electric_field_func(pos):
            # pos is a numpy array [x, y, z]
            x, y, _ = pos
            r_squared = x**2 + y**2

            # If inside the circle, the field is zero
            if r_squared < RADIUS**2:
                return np.array([0, 0, 0])

            # If at the origin, return a zero vector to avoid division by zero
            if r_squared == 0:
                return np.array([0, 0, 0])

            # Outside the circle, calculate the field:
            # E = E_uniform + E_dipole
            
            # 1. Uniform field pointing right
            E_uniform = np.array([0.5, 0, 0]) 
            
            # 2. Field from the induced dipole on the conductor
            k = (RADIUS**2) / (r_squared**2)
            E_dipole = k * np.array([
                (x**2 - y**2),
                (2 * x * y),
                0
            ])
            
            # The total field is the sum of the uniform field
            # MINUS the x-component of the dipole field but
            # PLUS the y-component. A simpler representation is:
            # E_total = E_uniform - E_dipole, if E_dipole's x was (y^2-x^2)
            # This is the correct formula for a conducting cylinder in a uniform E-field.
            return E_uniform - (0.5 * E_dipole)


        # --- Create the StreamLines ---
        stream_lines = StreamLines(
            electric_field_func,
            x_range=[-6, 6, 0.5],
            y_range=[-4, 4, 0.5],
            stroke_width=2,
            virtual_time=4, # Increase time to allow lines to curve around
            max_anchors_per_line=0.1, # Finer resolution
            color=BLUE
        )
        
        # --- Animate the scene ---
        self.play(FadeIn(circle))
        self.play(Create(stream_lines))
        self.wait()
        

# load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING ElectricFieldAroundCircle
```
