# Scene 1 Part 1
Create a Sheet with area vector A

```python
class gauss_s1p1(ThreeDScene):
    def construct(self):
        # Parameters
        grid_size = 3
        spacing = 2
        # arrow_height = 2
        # arrow_color = BLUE
        
        # Create the platform (prism)
        platform_length = spacing * grid_size * 1.05
        platform_width = spacing * grid_size * 1.05
        platform_height = 0.1
        platform_z = 0.05  # slight height above grid base

        platform = Prism(
            dimensions=[platform_length, platform_width, platform_height],
            fill_opacity=0.25,
            fill_color=GRAY,
            stroke_width=0.5
        )
        platform.move_to([0, 0, platform_z])

        # Create the normal vector (arrow) and label
        center = platform.get_center()
        normal_vector = Arrow3D(
            start=center,
            end=center + [0, 0, 1],  # Normal to flat surface = +z direction
            color=YELLOW,
            thickness=0.05
        )

        # Label for the vector A
        label = MathTex(r"\vec{A}", color=YELLOW).scale(0.8)
        label.next_to(normal_vector.get_end(), UP, buff=0.1).rotate(PI / 2, axis=RIGHT)

        # Group all together
        platform_group = VGroup(platform, normal_vector, label)
        self.add(platform_group)

        # Set camera slightly angled to see rotation
        self.set_camera_orientation(phi=75 * DEGREES, theta=-45 * DEGREES)
        self.wait(1)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING gauss_s1p1
```

# Scene 2 Part 2
Add the Flux Arrows at certain points

```python
class gauss_s2p2(Scene):
    def construct(self):
        # Create the main circle
        circle = Circle(radius=2, color=BLUE, fill_opacity=0.5)
        
        # Number of arrows around the circle
        num_arrows = 10
        radius = 2  # same as circle radius
        arrow_length = 0.8
        skip = [0, 5]
        dot_positions = []  # collect positions for electric field arrows

        arrows = VGroup()
        for i in range(num_arrows):
            if i in skip: # Skip certain indices if needed
                continue
            angle = i * TAU / num_arrows  # TAU = 2*pi
            # Start point on the circle
            start = radius * np.array([np.cos(angle), np.sin(angle), 0])
            # End point further out in same direction
            end = (radius + arrow_length) * np.array([np.cos(angle), np.sin(angle), 0])
            arrow = Arrow(start, end, buff=0, stroke_width=2, color=YELLOW)

            # Dot color logic
            dot_color = GREEN if 1 <= i <= 4 else RED # Setting the differnt dot
            dot = Dot(point=start, color=dot_color)
            dot_positions.append(start)  # save position for the electric field arrows

            # Group arrow and dot together
            pair = VGroup(arrow, dot)
            arrows.add(pair)
            
        self.play((circle))
        self.play(LaggedStart(*[FadeIn(pair) for pair in arrows], lag_ratio=0.1))
        self.wait()

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING gauss_s2p2
```

# Scene 4 Part 2
Arrows end points brought to center of the cirlce

```python
class gauss_s4p2(Scene):
    def construct(self):
        # Create the main circle
        circle = Circle(radius=2, color=BLUE, fill_opacity=0.5)
        
        # Number of arrows around the circle
        num_arrows = 8
        radius = 2  # same as circle radius
        arrow_length = 1.6

        arrows = VGroup()
        center = np.array([0, 0, 0])
        for i in range(num_arrows):
            angle = i * TAU / num_arrows  # TAU = 2*pi
            # All arrows start at the center of the circle
            start = radius * np.array([np.cos(angle), np.sin(angle), 0])
            # End point on the circle's edge
            end = (radius + arrow_length) * np.array([np.cos(angle), np.sin(angle), 0])
            arrow = Arrow(start, end, buff=0, stroke_width=2, color=YELLOW)
            arrows.add(arrow)

        self.add(circle)
        self.add(arrows)
        self.wait()

        # Create a new arrow group to hold the transformed arrows
        center_arrows = VGroup()

        # Change the end points of the arrows to the center of the circle
        for i in range(num_arrows):
            angle = i * TAU / num_arrows  # TAU = 2*pi
            # All arrows start at the center of the circle
            start = center
            new_end = (radius + arrow_length) * np.array([np.cos(angle), np.sin(angle), 0])
            center_arrow = Arrow(start, new_end, buff=0, stroke_width=2, color=YELLOW)
            center_arrows.add(center_arrow)

        # Drag the arrows to the center positions
        self.play(Transform(arrows, center_arrows), run_time=2)
        self.wait()

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING gauss_s4p2
```

# Scene 5 Part 1
Create postive charged circle with flux dots

```python
class gauss_s5p1(Scene):
    def construct(self):        
        num_arrows = 8
        radius = 2
        arrow_length = 1.6
        center = np.array([0, 0, 0])

        arrows = VGroup()
        angles = []

        for i in range(num_arrows):
            angle = i * TAU / num_arrows
            angles.append(angle)
            start = center
            end = (radius + arrow_length) * np.array([np.cos(angle), np.sin(angle), 0])
            arrow = Arrow(start, end, buff=0, stroke_width=2, color=YELLOW)
            arrows.add(arrow)

        # Positive charge
        positive_charge = Dot(point=center, color=RED).scale(2)
        plus_label = MathTex(r"+", color=WHITE).scale(0.5).move_to(center)

        # Create the main circle
        moving_circle = Circle(radius=radius, color=BLUE, fill_opacity=0.25)

        # Function to find the intersection of a ray from origin with moving circle
        def get_ray_circle_intersection(circle_center, radius, direction):
            direction = direction / np.linalg.norm(direction)
            oc = -circle_center
            a = 1
            b = 2 * np.dot(direction, oc)
            c = np.dot(oc, oc) - radius**2
            discriminant = b**2 - 4*a*c
            if discriminant < 0:
                return np.array([0, 0, 0])  # Fallback if no real intersection
            t = (-b + np.sqrt(discriminant)) / (2 * a)
            return t * direction

        # Create dynamically updating green dots at arrow-circle intersections
        dynamic_dots = VGroup()
        for angle in angles:
            direction = np.array([np.cos(angle), np.sin(angle), 0])
            def make_dot(direction=direction):
                def update():
                    pos = get_ray_circle_intersection(
                        moving_circle.get_center(),
                        radius,
                        direction
                    )
                    dist_to_center = np.linalg.norm(pos)
                    
                    # Map distance to radius: closer = bigger
                    r_max = 0.15
                    r_min = 0.05
                    d_max = radius + 1.5  # max expected distance for scaling
                    
                    radius_dot = r_max * (1 - min(dist_to_center / d_max, 1)) + r_min
                    return Dot(pos, color=WHITE, radius=radius_dot)
                return always_redraw(update)
            dynamic_dots.add(make_dot())

        # Add in everything at once
        self.play(
            AnimationGroup(
                FadeIn(moving_circle),
                FadeIn(arrows),
                FadeIn(positive_charge),
                FadeIn(plus_label),
                FadeIn(dynamic_dots),
                lag_ratio=0.0
            )
        )
        self.wait()


# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING gauss_s5p1
```

```python
class gauss_s5p2(Scene):
    def construct(self):        
        num_arrows = 8
        radius = 2
        arrow_length = 1.6
        center = np.array([0, 0, 0])

        arrows = VGroup()
        angles = []

        for i in range(num_arrows):
            angle = i * TAU / num_arrows
            angles.append(angle)
            start = center
            end = (radius + arrow_length) * np.array([np.cos(angle), np.sin(angle), 0])
            arrow = Arrow(start, end, buff=0, stroke_width=2, color=YELLOW)
            arrows.add(arrow)

        # Positive charge
        positive_charge = Dot(point=center, color=RED).scale(2)
        plus_label = MathTex(r"+", color=WHITE).scale(0.5).move_to(center)

        # Create the circle
        moving_circle = Circle(radius=radius, color=BLUE, fill_opacity=0.25)

        # Function to find the intersection of a ray from origin with moving circle
        def get_ray_circle_intersection(circle_center, radius, direction):
            direction = direction / np.linalg.norm(direction)
            oc = -circle_center
            a = 1
            b = 2 * np.dot(direction, oc)
            c = np.dot(oc, oc) - radius**2
            discriminant = b**2 - 4*a*c
            if discriminant < 0:
                return np.array([0, 0, 0])  # Fallback if no real intersection
            t = (-b + np.sqrt(discriminant)) / (2 * a)
            return t * direction

        # Create dynamically updating green dots at arrow-circle intersections
        dynamic_dots = VGroup()
        for angle in angles:
            direction = np.array([np.cos(angle), np.sin(angle), 0])
            def make_dot(direction=direction):
                def update():
                    pos = get_ray_circle_intersection(
                        moving_circle.get_center(),
                        radius,
                        direction
                    )
                    dist_to_center = np.linalg.norm(pos)
                    
                    # Map distance to radius: closer = bigger
                    r_max = 0.15
                    r_min = 0.05
                    d_max = radius + 1.5  # max expected distance for scaling
                    
                    radius_dot = r_max * (1 - min(dist_to_center / d_max, 1)) + r_min
                    return Dot(pos, color=WHITE, radius=radius_dot)
                return always_redraw(update)
            dynamic_dots.add(make_dot())

        # Add in everything at once
        self.play(
            AnimationGroup(
                FadeIn(moving_circle),
                FadeIn(arrows),
                FadeIn(positive_charge),
                FadeIn(plus_label),
                FadeIn(dynamic_dots),
                lag_ratio=0.0
            )
        )
        self.wait()

        # Move the circle left and then right while dots update in real time
        self.play(moving_circle.animate.shift(LEFT * 1), run_time=5) # Move left
        self.wait()
        self.play(moving_circle.animate.shift(RIGHT * 2), run_time=5) # Move right
        self.wait()
        self.play(moving_circle.animate.shift(LEFT * 1), run_time=5) # Return to center
        self.wait()


# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING gauss_s5p2
```

# Scene 6 Part 2
Turn the relation into an equation 

```python
class gauss_s6p2(Scene):
    def construct(self):
        # Step 1: Show standalone Phi and Q relation
        initial_equation = MathTex(r"\Phi_{\text{net}} \propto Q_{\text{in}}", font_size=72)

        self.add(initial_equation)
        self.wait()

        # Step 2: Transform into the equation
        updated_equation = MathTex(r"\Phi_{\text{net}} = \frac{1}{\epsilon_0} Q_{\text{in}}", font_size=72)
        updated_equation.move_to(initial_equation.get_center())

        self.play(ReplacementTransform(initial_equation, updated_equation))
        self.wait(2)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING gauss_s6p2
```

# Scene 6 Part 3
Turn into the Gauss Law Equation

```python
class gauss_s6p3(Scene):
    def construct(self):
        # Step 1: Show standalone Phi and Q relation
        initial_equation = MathTex(r"\Phi_{\text{net}} \propto Q_{\text{in}}", font_size=72)

        # # Step 2: Transform into the equation
        updated_equation = MathTex(r"\Phi_{\text{net}} = \frac{1}{\epsilon_0} Q_{\text{in}}", font_size=72)

        # Add the previous equation directly
        self.add(updated_equation)

        # Step 3: Integration :)
        updated_equationIntegral = MathTex(
            r"\oint \vec{E} \cdot d\vec{A} = \frac{1}{\epsilon_0} Q_{\text{in}}",
            font_size=72
        )
        updated_equationIntegral.move_to(initial_equation.get_center())

        self.play(ReplacementTransform(updated_equation, updated_equationIntegral))
        self.wait(2)


# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING gauss_s6p3
```

# Scene 7 Part 1
Tranform the Circle into a Random Gaussian Surface

```python
import random # This import is necessary for the random number generation in the blob shape

class gauss_s7p1(MovingCameraScene):
    def construct(self):        
        num_arrows = 8
        radius = 2
        arrow_length = 1.6
        center = np.array([0, 0, 0])

        arrows = VGroup()
        angles = []

        for i in range(num_arrows):
            angle = i * TAU / num_arrows
            angles.append(angle)
            start = center
            end = (radius + arrow_length) * np.array([np.cos(angle), np.sin(angle), 0])
            arrow = Arrow(start, end, buff=0, stroke_width=2, color=YELLOW)
            arrows.add(arrow)

        # Positive charge
        positive_charge = Dot(point=center, color=RED).scale(2)
        plus_label = MathTex(r"+", color=WHITE).scale(0.5).move_to(center)

        # Moving circle
        moving_circle = Circle(radius=radius, color=BLUE, fill_opacity=0.25)

        # All appear together
        self.play(
            AnimationGroup(
                FadeIn(moving_circle),
                FadeIn(arrows),
                FadeIn(positive_charge),
                FadeIn(plus_label),
                lag_ratio=0.0
            )
        )
        self.wait()

        # Animate the camera slowly zooming out
        self.play(self.camera.frame.animate.scale(1.5), run_time=3)

        # Create a Random Gaussian surface
        def generate_organic_blob(center, base_radius, num_points=100, amplitude=0.4, frequency=3):
            points = []
            for i in range(num_points):
                angle = i * TAU / num_points
                r = base_radius + amplitude * np.sin(frequency * angle + random.uniform(0, TAU))
                x = r * np.cos(angle)
                y = r * np.sin(angle)
                points.append([x, y, 0])
            
            # Create a VMobject with smooth curve through these points
            blob = VMobject(color=BLUE, fill_opacity=0.25)
            blob.set_points_smoothly([*points, points[0]])  # close the loop
            blob.move_to(center)
            return blob

        # Generate the blob shape centered on the moving circle
        blob_shape = generate_organic_blob(
            center=moving_circle.get_center(),
            base_radius=random.uniform(2.5, 3.5),  # Random base radius for variety
            num_points=random.randint(80, 120),  # Random number of points for variety
            amplitude=random.uniform(0.2, 0.6),  # Random amplitude for variety
            frequency=random.randint(5, 15)  # Random frequency for variety
        )

        # Print the values used for the blob shape if they end up being useful
        print(f"Blob shape parameters: base_radius={blob_shape.get_width()/2}, num_points={len(blob_shape.get_points())}, amplitude={blob_shape.get_height()/2}, frequency={blob_shape.get_num_curves()}")


        # Match styling
        blob_shape.set_fill(BLUE, opacity=0.25)

        # Animate transformation from circle to organic shape
        self.play(Transform(moving_circle, blob_shape), run_time=2)
        self.wait()



# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING gauss_s7p1
```
