```python

class s1p1(Scene):
    def construct(self):
        number_plane = NumberPlane(
            background_line_style={
                "stroke_color": BLUE_D,
                "stroke_width": 4,
                "stroke_opacity": 0.2,

            }
        )# Axis labels
        x_label = MathTex("x").next_to(number_plane.c2p(6, 0), RIGHT*1.5+DOWN*0.7)
        y_label = MathTex("y").next_to(number_plane.c2p(0, 3), UP+RIGHT*0.7+UP*0.2)
        grass = SVGMobject("../../images/grass.svg").scale(4)
        dot = Dot(color=YELLOW, radius=0.2).move_to(number_plane.c2p(0, 0))
        dot_border = Dot(color=BLACK, radius=0.22).move_to(number_plane.c2p(0, 0))

        test_dot = VGroup(dot_border, dot)

        self.play(FadeIn(grass))

        line1 = Line(start=number_plane.c2p(-1, 1), end=number_plane.c2p(1, -1), color=BLACK, stroke_width=2).scale(0.15)
        line2 = Line(start=number_plane.c2p(1, 1), end=number_plane.c2p(-1, -1), color=BLACK, stroke_width=2).scale(0.15)
        starting_point = VGroup(line1, line2)

        self.play(GrowFromCenter(test_dot))
        self.wait(1)

        compass_directions = VGroup()
        north = MathTex("N").next_to(dot, UP).scale(0.7)
        south = MathTex("S").next_to(dot, DOWN).scale(0.7)
        east = MathTex("E").next_to(dot, RIGHT).scale(0.7)
        west = MathTex("W").next_to(dot, LEFT).scale(0.7)
        n_arrow = Arrow(dot.get_center(), dot.get_center() + UP*0.5, buff=0, color=RED, stroke_width=6)
        s_arrow = Arrow(dot.get_center(), dot.get_center() + DOWN*0.5, buff=0, color=RED, stroke_width=6)
        e_arrow = Arrow(dot.get_center(), dot.get_center() + RIGHT*0.5, buff=0, color=RED, stroke_width=6)
        w_arrow = Arrow(dot.get_center(), dot.get_center() + LEFT*0.5, buff=0, color=RED, stroke_width=6)
        compass_directions.add(n_arrow, s_arrow, e_arrow, w_arrow)
        compass_directions.add(north, south, east, west)

        compass_directions.move_to(LEFT*6 + UP *3)
        self.play(Write(compass_directions))
        self.wait(2)
        dashed_1 = DashedLine(start=dot.get_center(), end=dot.get_center() + RIGHT*2, color=RED, stroke_width=3)
        dashed_lab_1 = MathTex(r"d_1").next_to(dashed_1, UP).set_color(RED)
        dashed_2 = DashedLine(start=dot.get_center() + RIGHT*2, end=dot.get_center() + RIGHT*5, color=BLUE, stroke_width=3)
        dashed_lab_2 = MathTex(r"d_2").next_to(dashed_2, UP).set_color(BLUE)


        self.play(test_dot.animate.move_to(number_plane.c2p(2, 0)), FadeIn(starting_point), GrowFromEdge(dashed_1, LEFT), Write(dashed_lab_1))
        self.wait(2)
        self.play(test_dot.animate.move_to(number_plane.c2p(5, 0)), GrowFromEdge(dashed_2, LEFT), Write(dashed_lab_2))
        self.wait(2)

        dashed_lab_tot = MathTex(r"d_T", color=YELLOW).next_to(dashed_1.get_end(), DOWN+RIGHT)
        self.play(Write(dashed_lab_tot))
        self.wait(2)

        eq_1 = MathTex(r"d_T", "=", r"d_1", "+", r"d_2").to_edge(UP).scale(1.5)
        eq_1[0].set_color(YELLOW)
        eq_1[2].set_color(RED)
        eq_1[4].set_color(BLUE)
        self.play(TransformFromCopy(dashed_lab_tot, eq_1[0]), Write(eq_1[1]))
        #self.play(Write(eq_1[0:2]))
        self.wait(2)

        self.play(TransformFromCopy(dashed_lab_1, eq_1[2]), TransformFromCopy(dashed_lab_2, eq_1[4]), Write(eq_1[3]))
        self.wait(2)

        dashed_1_value = MathTex(r"2 m").set_color(RED).move_to(dashed_lab_1)
        dashed_2_value = MathTex(r"3 m").set_color(BLUE).move_to(dashed_lab_2)
        dashed_tot_value = MathTex(r"5 m").set_color(YELLOW).move_to(dashed_lab_tot)

        self.play(ReplacementTransform(dashed_lab_1, dashed_1_value), Transform(eq_1[2], dashed_1_value.copy().move_to(eq_1[2]).scale(1.2)))
        self.wait(1)
        self.play(ReplacementTransform(dashed_lab_2, dashed_2_value), Transform(eq_1[4], dashed_2_value.copy().move_to(eq_1[4]).scale(1.2)))
        self.wait(1)
        self.play(ReplacementTransform(dashed_lab_tot, dashed_tot_value), Transform(eq_1[0], dashed_tot_value.copy().move_to(eq_1[0]).scale(1.2)))
        self.wait(2)

        self.play(dashed_1.animate.shift(RIGHT*3), dashed_2.animate.shift(LEFT*2), dashed_1_value.animate.shift(RIGHT*3), dashed_2_value.animate.shift(LEFT*2),
                  eq_1[2].animate.move_to(eq_1[4]), eq_1[4].animate.move_to(eq_1[2]))

        self.wait(1)


        self.play(test_dot.animate.move_to(number_plane.c2p(0, 0)), 
                  FadeOut(dashed_1), FadeOut(dashed_2), 
                  FadeOut(dashed_1_value), FadeOut(dashed_2_value), FadeOut(dashed_tot_value), 
                  FadeOut(eq_1), FadeOut(starting_point),
                  FadeOut(dashed_1_value.copy()), FadeOut(dashed_2_value.copy()), FadeOut(dashed_tot_value.copy())
        )

        self.wait(2)
         # Add the number plane and labels to the scene


        #self.play(Create(number_plane))
        #self.wait(1)
        


# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING s1p1
```

```python
class s1p3(Scene):
    def construct(self):
        number_plane = NumberPlane(
            background_line_style={
                "stroke_color": BLUE_D,
                "stroke_width": 4,
                "stroke_opacity": 0.2,

            }
        )# Axis labels
        x_label = MathTex("x").next_to(number_plane.c2p(6, 0), RIGHT*1.5+DOWN*0.7)
        y_label = MathTex("y").next_to(number_plane.c2p(0, 3), UP+RIGHT*0.7+UP*0.2)


        # Starting Setup
        grass = SVGMobject("../../images/grass.svg").scale(4)
        dot = Dot(color=YELLOW, radius=0.2).move_to(number_plane.c2p(0, 0))
        dot_border = Dot(color=BLACK, radius=0.22).move_to(number_plane.c2p(0, 0))
        test_dot = VGroup(dot_border, dot)
        line1 = Line(start=number_plane.c2p(-1, 1), end=number_plane.c2p(1, -1), color=BLACK, stroke_width=2).scale(0.15)
        line2 = Line(start=number_plane.c2p(1, 1), end=number_plane.c2p(-1, -1), color=BLACK, stroke_width=2).scale(0.15)
        starting_point = VGroup(line1, line2)

        compass_directions = VGroup()
        north = MathTex("N").next_to(dot, UP).scale(0.7)
        south = MathTex("S").next_to(dot, DOWN).scale(0.7)
        east = MathTex("E").next_to(dot, RIGHT).scale(0.7)
        west = MathTex("W").next_to(dot, LEFT).scale(0.7)

        n_arrow = Arrow(dot.get_center(), dot.get_center() + UP*0.5, buff=0, color=RED, stroke_width=6)
        s_arrow = Arrow(dot.get_center(), dot.get_center() + DOWN*0.5, buff=0, color=RED, stroke_width=6)
        e_arrow = Arrow(dot.get_center(), dot.get_center() + RIGHT*0.5, buff=0, color=RED, stroke_width=6)
        w_arrow = Arrow(dot.get_center(), dot.get_center() + LEFT*0.5, buff=0, color=RED, stroke_width=6)

        compass_directions.add(n_arrow, s_arrow, e_arrow, w_arrow)
        compass_directions.add(north, south, east, west)

        eq_1 = MathTex(r"d_T", "=", r"d_1", "+", r"d_2").to_edge(UP).scale(1.5)
        eq_1[0].set_color(YELLOW)
        eq_1[2].set_color(RED)
        eq_1[4].set_color(BLUE)

        compass_directions.move_to(LEFT*6 + UP *3)
        self.add(grass, compass_directions, test_dot)
        self.wait(1)

        step_1 = 3
        step_2 = 2

        # Scene part specific

        d1 = Arrow(start=ORIGIN, end=RIGHT*step_1, color=RED, stroke_width=4, buff=0)
        d2 = Arrow(start=d1.get_end(), end=d1.get_end() + UP*step_2, color=BLUE, stroke_width=4, buff=0)
        d3 = Arrow(start=ORIGIN, end=RIGHT*step_1 + UP*step_2, color=YELLOW, stroke_width=4, buff=0)

        d1_label = MathTex(r"d_1").set_color(RED).next_to(d1.get_center(), DOWN+RIGHT*0.2)
        d2_label = MathTex(r"d_2").set_color(BLUE).next_to(d2.get_center(), RIGHT*0.8)
        d3_label = MathTex(r"d_T").set_color(YELLOW).next_to(d3.get_center(), UP)

        self.play(test_dot.animate.move_to(number_plane.c2p(step_1, 0)), FadeIn(starting_point), GrowArrow(d1), Write(d1_label))
        self.wait(1)
        self.play(test_dot.animate.move_to(number_plane.c2p(step_1, step_2)), GrowArrow(d2), Write(d2_label))
        self.wait(1)
        self.play(GrowArrow(d3), Write(d3_label))
        self.wait(2)

        self.play(TransformFromCopy(d3_label, eq_1[0]), Write(eq_1[1]))
        #self.play(Write(eq_1[0:2]))
        self.wait(2)

        self.play(TransformFromCopy(d1_label, eq_1[2]), TransformFromCopy(d2_label, eq_1[4]), Write(eq_1[3]))
        self.wait(2)

        eq_2 = MathTex(
            r"d_T",
            "=",
            r"\sqrt{",  # Square root symbol
            r"d_1^2",   # First term
            r"+",       # Plus sign
            r"d_2^2",   # Second term
            r"}"        # Closing brace
        ).to_edge(UP).scale(1.5)

        # Set colors for the terms
        eq_2[0].set_color(YELLOW)  # Square root in yellow
        eq_2[3].set_color(RED)  # d1^2 in red
        eq_2[5].set_color(BLUE) # d2^2 in blue

        self.play(ReplacementTransform(eq_1, eq_2))

        self.wait(2)

        pyth_txt = Text("Pythagorean Theorem").move_to(DOWN*2)
        pythagg = MathTex(r"a^2", "+", r"b^2", "=", r"c^2").next_to(pyth_txt, DOWN)
        pythagg[0].set_color(RED)
        pythagg[2].set_color(BLUE)
        pythagg[4].set_color(YELLOW)
        self.play(Write(pyth_txt), Write(pythagg))
        self.wait(2)

        d1_val = MathTex(r"3 m").set_color(RED).move_to(d1_label)
        d2_val = MathTex(r"4 m").set_color(BLUE).move_to(d2_label)
        d3_val = MathTex(r"5 m").set_color(YELLOW).move_to(d3_label)

        self.play(ReplacementTransform(d1_label, d1_val), ReplacementTransform(d2_label, d2_val),
                  Transform(eq_2[3], MathTex(r"3^2").move_to(eq_2[3]).scale(1.2).set_color(RED)), 
                  Transform(eq_2[5], MathTex(r"4^2").move_to(eq_2[5]).scale(1.2).set_color(BLUE)))
        self.wait(1)

        self.play(Transform(eq_2[3], MathTex(r"9").move_to(eq_2[3]).scale(1.2).set_color(RED)), 
                  Transform(eq_2[5], MathTex(r"16").move_to(eq_2[5]).scale(1.2).set_color(BLUE)))
        self.wait(2)

        eq_3 = MathTex(
            r"d_T",
            "=",
            r"\sqrt{",
            r"25",
            r"}"
        ).to_edge(UP).scale(1.5)

        eq_3[0].set_color(YELLOW)
        eq_3[3].set_color(YELLOW)
        self.play(ReplacementTransform(eq_2, eq_3))
        self.wait(2)

        eq_4 = MathTex(
            r"d_T",
            "=",
            r"5 m"
        ).to_edge(UP).scale(1.5)

        eq_4[0].set_color(YELLOW)
        eq_4[2].set_color(YELLOW)


        self.play(ReplacementTransform(d3_label, d3_val), ReplacementTransform(eq_3, eq_4))
        self.wait(2)


        

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING s1p3
```

```python
class s2p1(Scene):
    def construct(self):
        number_plane = NumberPlane(
            background_line_style={
                "stroke_color": BLUE_D,
                "stroke_width": 4,
                "stroke_opacity": 0.2,

            }
        )# Axis labels
        x_label = MathTex("x").next_to(number_plane.c2p(6, 0), RIGHT*1.5+DOWN*0.7)
        y_label = MathTex("y").next_to(number_plane.c2p(0, 3), UP+RIGHT*0.7+UP*0.2)


        # Starting Setup

        dot = Dot(color=PURE_GREEN, radius=0.1).move_to(number_plane.c2p(0, 0))


        # Scene part specific

        self.play(Create(number_plane), Write(x_label), Write(y_label))
        self.wait(1)
        self.play(Indicate(x_label, scale_factor=1.5))
        self.wait(1)
        self.play(Indicate(y_label, scale_factor=1.5))
        self.wait(1)

        self.play(GrowFromCenter(dot))
        self.wait(1)


        # Create brackets that display the x-coordinate of the dot
        brackets = always_redraw(
            lambda: MathTex(
                f"[{dot.get_x():.2f}]"
            ).next_to(dot, DOWN)
        )
        self.play(Write(brackets))

        box = Rectangle(
            width=5, height=2,
            fill_color=BLACK, fill_opacity=1,
            stroke_color=WHITE, stroke_width=4
        )
        box.move_to(LEFT*4 + UP*2)  # or wherever you want it

        # Create your MathTex
        eqn = always_redraw(
            lambda: MathTex(
                        r"\vec{d}", 
                        "=", 
                        f"{dot.get_x():.2f}", 
                        r"\hat{x}", 
                        color=WHITE
                        ).move_to(box.get_center()).scale_to_fit_width(box.width * 0.65)
            )
        

        # Animate the dot moving to 3,0
        self.play(dot.animate.move_to(number_plane.c2p(3, 0)), run_time=5)
        self.wait(1)
        arrow = always_redraw(lambda:Arrow(start=number_plane.c2p(0, 0), end=dot.get_center(), color=YELLOW, stroke_width=4, buff=0))
        arrow_1_label = always_redraw(lambda: MathTex(r"\vec{d}").set_color(WHITE).next_to(arrow.get_center(), UP))
        self.play(GrowArrow(arrow), Write(arrow_1_label))
        self.wait(2)
        self.play(DrawBorderThenFill(box))
        self.play(TransformFromCopy(arrow_1_label, eqn[0]))
        self.play(Write(eqn))
        self.wait(1)


        # Move dot back to origin
        self.play(dot.animate.move_to(number_plane.c2p(0, 0)), run_time=5)
        self.wait(1)


        self.play(dot.animate.move_to(number_plane.c2p(-2, 0)), run_time=5)
        self.wait(1)


        self.play(dot.animate.move_to(number_plane.c2p(0, 0)), run_time=5)
        self.wait(2)


        # Create brackets that display the x-coordinate of the dot
        brackets_2 = always_redraw(
            lambda: MathTex(
                f"[{dot.get_x():.2f}, {dot.get_y():.2f}]"
            ).next_to(dot, UP)
        )

        # Create your MathTex
        eqn_2 = always_redraw(
            lambda: MathTex(
                        r"\vec{d}", 
                        "=", 
                        f"{dot.get_x():.2f}", 
                        r"\hat{x}",
                        "+",
                        f"{dot.get_y():.2f}", 
                        r"\hat{y}", 
                        color=WHITE
                        ).move_to(box.get_center()).scale_to_fit_width(box.width * 0.65)
            )
        
        self.play(ReplacementTransform(brackets, brackets_2), ReplacementTransform(eqn, eqn_2))
        self.wait(2)
        self.play(dot.animate.move_to(number_plane.c2p(3, 0)), run_time=5)
        self.wait(1)
        self.play(dot.animate.move_to(number_plane.c2p(3, 2)), run_time=5)
        self.wait(1)

        x_dir = Arrow(start=number_plane.c2p(0, 0), end=number_plane.c2p(3, 0), color=RED, stroke_width=4, buff=0)
        x_lab = MathTex(r"d_x").set_color(RED).next_to(x_dir.get_center(), DOWN)
        y_dir = Arrow(start=number_plane.c2p(3, 0), end=number_plane.c2p(3, 2), color=BLUE, stroke_width=4, buff=0)
        y_lab = MathTex(r"d_y").set_color(BLUE).next_to(y_dir.get_center(), RIGHT)

        self.play(GrowArrow(x_dir), Write(x_lab), eqn_2[2:4].animate.set_color(RED))
        self.wait(1)
        self.play(GrowArrow(y_dir), Write(y_lab), eqn_2[5:].animate.set_color(BLUE))
        self.wait(2)
        

        

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING s2p1
```

```python
class open_vector(Scene):
    def construct(self):
        number_plane = NumberPlane(
            background_line_style={
                "stroke_width": 1,
                "stroke_opacity": 0.6
            }
        )# Axis labels
        x_label = MathTex("x").next_to(number_plane.c2p(6, 0), RIGHT+DOWN*0.5)
        y_label = MathTex("y").next_to(number_plane.c2p(0, 3), UP+RIGHT*0.7+UP*0.2)
        self.bring_to_back(number_plane)
        self.play(Create(number_plane), Write(x_label), Write(y_label))
        self.wait(1)

        arrow1 = Arrow(start=ORIGIN, end=2*RIGHT + UP, color=BLUE, buff=0)
        arrow2 = Arrow(start=ORIGIN, end=RIGHT + 2*UP, color=GREEN, buff=0)
        arrow3 = Arrow(start=ORIGIN, end=3*LEFT + 1*DOWN, color=PURE_RED, buff=0)

        label_1 = MathTex("\\vec{E}_1", color = BLUE).next_to(arrow1.get_end())
        label_2 = MathTex("\\vec{E}_2", color = GREEN).next_to(arrow2.get_end(), UP)
        label_3 = MathTex("\\vec{E}_3", color = PURE_RED).next_to(arrow3.get_end(), DOWN)

        self.play(GrowArrow(arrow1), GrowArrow(arrow2), GrowArrow(arrow3))
        self.play(Write(label_1), Write(label_2), Write(label_3))
        self.wait(2)



        box = Rectangle(
            width=5, height=3,
            fill_color=BLACK, fill_opacity=1,
            stroke_color=WHITE, stroke_width=4
        )
        box.move_to(LEFT*3.5 + UP*2)  # or wherever you want it

        # Create your MathTex
        eqn = MathTex(r"\vec{E}_{net}", "=", r"\vec{E}_1", "+", r"\vec{E}_2", "+", r"\vec{E}_3", color=WHITE)
        eqn[2].set_color(BLUE)
        eqn[4].set_color(GREEN)
        eqn[6].set_color(PURE_RED)
        eqn.scale_to_fit_width(box.width * 0.8)  # 80% of box width
        eqn.move_to(box.get_center())

        self.play(FadeIn(box), Write(eqn))
        self.wait(2)

        self.play(arrow2.animate.shift(RIGHT*2 + UP*1), label_2.animate.shift(RIGHT*2 + UP*1))
        self.play(arrow3.animate.shift(RIGHT*3 + UP*3), label_3.animate.shift(RIGHT*3.7 + UP*4.5))
        self.wait(1)

        arrow4 = Arrow(start = ORIGIN, end = UP*2, color=YELLOW, buff=0)
        label_4 = MathTex("\\vec{E}_{net}", color = YELLOW).next_to(arrow4.get_end(), RIGHT+DOWN*1)
        self.play(GrowArrow(arrow4), Write(label_4), eqn[0].animate.set_color(YELLOW))
        self.wait(2)


        self.play(arrow2.animate.shift(LEFT*2 + DOWN*1), label_2.animate.shift(LEFT*2 + DOWN*1),
            arrow3.animate.shift(LEFT*3 + DOWN*3), label_3.animate.shift(LEFT*3.7 + DOWN*4.5),
            FadeOut(arrow4), FadeOut(label_4), eqn[0].animate.set_color(WHITE))
        self.wait(2)


        e1x = Arrow(start=ORIGIN, end=2*RIGHT, color=BLUE, buff=0)
        e1y = Arrow(start=RIGHT*2, end=UP+RIGHT*2, color=BLUE, buff=0)
        e1x_lab = MathTex(r"\vec{E}_{1x}", color=BLUE).next_to(e1x.get_end(), LEFT*1.7+DOWN*0.4).scale(0.7)
        e1y_lab = MathTex(r"\vec{E}_{1y}", color=BLUE).next_to(e1y.get_end(), RIGHT+DOWN*1).scale(0.7)

        self.play(ReplacementTransform(arrow1.copy(), e1x), ReplacementTransform(arrow1.copy(), e1y), 
                  arrow1.animate.set_opacity(0.2), label_1.animate.set_opacity(0), 
                  Write(e1x_lab), Write(e1y_lab))
        self.wait(1)


        e2x = Arrow(start=UP*2, end=UP*2 + RIGHT, color=GREEN, buff=0)
        e2y = Arrow(start=ORIGIN, end=2*UP, color=GREEN, buff=0)
        e2x_lab = MathTex(r"\vec{E}_{2x}", color=GREEN).next_to(e2y.get_end(), RIGHT*.6+UP*.6).scale(0.7)
        e2y_lab = MathTex(r"\vec{E}_{2y}", color=GREEN).next_to(e2x.get_end(), DOWN*2+LEFT*4.5).scale(0.7)
        self.play(ReplacementTransform(arrow2.copy(), e2x), ReplacementTransform(arrow2.copy(), e2y), 
                  arrow2.animate.set_opacity(0.2), label_2.animate.set_opacity(0), 
                  Write(e2x_lab), Write(e2y_lab))
        self.wait(1)

        e3x = Arrow(start=ORIGIN, end=3*LEFT, color=PURE_RED, buff=0)
        e3y = Arrow(start=LEFT*3, end=LEFT*3+1*DOWN, color=PURE_RED, buff=0)
        e3x_lab = MathTex(r"\vec{E}_{3x}", color=PURE_RED).next_to(e3x.get_end(), RIGHT*3.7+DOWN*0.05).scale(0.7)
        e3y_lab = MathTex(r"\vec{E}_{3y}", color=PURE_RED).next_to(e3y.get_end(), UP*1+LEFT*0.5).scale(0.7)
        self.play(ReplacementTransform(arrow3.copy(), e3x), ReplacementTransform(arrow3.copy(), e3y), 
                  arrow3.animate.set_opacity(0.2), label_3.animate.set_opacity(0), 
                  Write(e3x_lab), Write(e3y_lab))
        self.wait(2) 


        self.play(e2x.animate.shift(RIGHT*2 + DOWN*2), e2x_lab.animate.shift(RIGHT*2 + DOWN*2.8),
                  e1y.animate.shift(UP*2 + LEFT*2), e1y_lab.animate.shift(UP*2 + LEFT*3.1),
                  e3y.animate.shift(RIGHT*3), e3y_lab.animate.shift(RIGHT*3),
                  arrow1.animate.set_opacity(0), arrow2.animate.set_opacity(0), arrow3.animate.set_opacity(0))
        self.wait(1)


        eqn2 = MathTex(r"\vec{E}_{net,x}", "=", r"\vec{E}_{1x}", "+", r"\vec{E}_{2x}", "+", r"\vec{E}_{3x} \\ \\", "\\",
                       r"\vec{E}_{net,y}", "=", r"\vec{E}_{1y}", "+", r"\vec{E}_{2y}", "+", r"\vec{E}_{3y}", color=WHITE)
        
        eqn2[2].set_color(BLUE)
        eqn2[4].set_color(GREEN)
        eqn2[6].set_color(PURE_RED)
        eqn2[10].set_color(BLUE)
        eqn2[12].set_color(GREEN)
        eqn2[14].set_color(PURE_RED)

        eqn2.scale_to_fit_width(box.width * 0.8)  # 80% of box width
        eqn2.move_to(box.get_center())
        self.play(ReplacementTransform(eqn, eqn2))
        self.wait(2)

        xgroup = VGroup(e1x, e2x, e3x, e1x_lab, e2x_lab, e3x_lab)
        ygroup = VGroup(e1y, e2y, e3y, e1y_lab, e2y_lab, e3y_lab)

        self.add(xgroup, ygroup)

        e_net_x = Arrow(start=ORIGIN, end= (2+1)*RIGHT + 3*LEFT, color=YELLOW, buff=0)
        e_net_y = Arrow(start=ORIGIN, end= (1+2)*UP + 1*DOWN, color=YELLOW, buff=0)

        eqn3 = MathTex(r"\vec{E}_{net} = \vec{E}_{net,x} + \vec{E}_{net,y}", color=YELLOW)
        eqn3.scale_to_fit_width(box.width * 0.8)  # 80% of box width
        eqn3.move_to(box.get_center())

        self.play(Transform(xgroup, e_net_x))

        self.play(Transform(ygroup, e_net_y),
                  Write(label_4), ReplacementTransform(eqn2, eqn3))

        self.wait(1)



# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING open_vector
```
