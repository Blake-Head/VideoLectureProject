```python
from manim import *
import numpy as np
import random
import manim_physics as mp
```

# Scene 1

## (s1p1)

```python
class s1p1(Scene):
    def construct(self):
        # Create the charges and set the position
        charge1 = mp.Charge(magnitude=4, point=np.array([-4, -1, 0]))
        charge2 = mp.Charge(magnitude=3, point=np.array([4, 1, 0]))
        charge3 = mp.Charge(magnitude=-3, point=np.array([0, 3, 0]))


        efield_1 = mp.ElectricField(charge1)
        efield_2 = mp.ElectricField(charge2)
        efield_3 = mp.ElectricField(charge3)

        total_efield = mp.ElectricField(charge1, charge2, charge3, color=YELLOW)

        self.play(FadeIn(charge1), FadeIn(charge2), FadeIn(charge3))
        self.wait(1)
        self.play(FadeIn(efield_1), FadeIn(efield_2), FadeIn(efield_3))
        self.wait(2)

        self.play(FadeOut(efield_1, efield_2, efield_3), FadeIn(total_efield))
        self.wait(2)




# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING s1p1
```

## (s1p2)

```python
class s1p2(Scene):
    def construct(self):
        # Create the charge and its field
        charge_pos = mp.Charge(magnitude=2)
        self.play(FadeIn(charge_pos))

        # Parameters for radial lines
        num_lines = 18  # 16 looks good
        line_length = 8  # Adjust as needed to go beyond scene edge

        # Center of charge
        center = charge_pos.get_center()

        # Create and add lines
        lines = VGroup()
        for i in range(num_lines):
            angle = i * TAU / num_lines  # TAU = 2*pi
            direction = np.array([np.cos(angle), np.sin(angle), 0])
            start = center
            end = center + direction * line_length
            line = Line(start, end, color=YELLOW, stroke_width=2)
            arrow = Arrow(start, end/(0.35*line_length), color=YELLOW, buff=0, stroke_width=1)
            lines.add(line)
            lines.add(arrow)

        self.play(FadeIn(lines))
        self.wait(1)


        box = Rectangle(
            width=4, height=2.5,
            fill_color=BLACK, fill_opacity=1,
            stroke_color=WHITE, stroke_width=4
        )
        box.move_to(LEFT*4.5 + UP*2)  # or wherever you want it

        eqn = MathTex(
            r"\vec{E} = \frac{kQ}{", "r", r"^2}", r"\hat{r}",
            color=WHITE
        )
        # Example: color r in red
        #eqn[1].set_color(RED)
        

        eqn.scale_to_fit_width(box.width * 0.8)  # 80% of box width
        eqn.move_to(box.get_center())

        self.play(FadeIn(box), Write(eqn))
        self.wait(2)
        end_point = Dot(np.array([3, 2, 0]), fill_opacity=0)

        r_arrow = always_redraw(lambda: Arrow(
                        start=ORIGIN, 
                        end=end_point.get_center(), 
                        color=PURE_GREEN, 
                        buff=0,
                        stroke_width=2))
        
        r_label = always_redraw(lambda: MathTex(r"r", 
                                                color=PURE_GREEN).next_to(r_arrow.get_center(), DOWN))


        self.play(ShrinkToCenter(charge_pos), run_time=5)
        self.wait(1)
        self.play(eqn[1].animate.set_color(PURE_GREEN), GrowFromPoint(r_arrow, np.array([0, 0, 0])), Write(r_label))
        self.wait(1)

        self.play(end_point.animate.move_to(np.array([5, 3, 0])), run_time=3)
        self.wait(2)

        self.play(end_point.animate.move_to(np.array([3.5, 0, 0])), run_time=3)
        self.wait(2)

        circ_path = Circle(radius=3.5, color=PURE_BLUE).move_to(charge_pos.get_center())
        self.play(MoveAlongPath(end_point, circ_path), run_time=6)
        self.wait(2)

        charge_pos_2 = mp.Charge(magnitude=1, add_glow=False).move_to(np.array([0, 0, 0]))
        self.play(GrowFromCenter(charge_pos_2), FadeOut(box), FadeOut(eqn), FadeOut(r_arrow), FadeOut(r_label), FadeOut(end_point))
        self.wait(1)
        self.play(FadeOut(lines))
        self.wait(1)



# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING s1p2
```

# Scene 2

## (s2p1)

```python
class s2p1(Scene):
    def construct(self):

        charge = mp.Charge(magnitude=1, add_glow=False)
        charge_lab = MathTex("Q").next_to(charge, RIGHT)
        self.add(charge)
        self.play(Write(charge_lab))
        self.wait(2)

        linecharge = Line(start=LEFT*5, end=RIGHT*5, color=RED, stroke_width=8).shift(DOWN*2)
        self.play(ReplacementTransform(charge, linecharge), charge_lab.animate.next_to(linecharge, DOWN))
        self.wait(2)

        moving_dot = Dot(point=linecharge.get_start(), color=YELLOW)
        self.play(FadeIn(moving_dot))
        self.wait(1) 
        self.play(MoveAlongPath(moving_dot, linecharge, rate_func=rate_functions.ease_in_out_quart, run_time=5))
        self.play(FadeOut(moving_dot))
        self.wait(1)

        random_dots = VGroup()
        random_labs = VGroup()
        num_dots = 20
        for _ in range(num_dots):
            x = random.uniform(-5, 5)
            y = -2
            dot = Dot(radius = 0.05, point=np.array([x, y, 0]), color=YELLOW)
            lab = MathTex("dq").next_to(dot, UP*0.3).scale(0.35)
            random_labs.add(lab)
            random_dots.add(dot)
            
        self.play(FadeIn(random_dots), Write(random_labs))
        self.wait(2)
        self.play(FadeOut(random_dots, random_labs))
        self.wait(1)

        poi = Dot(point=np.array([0, 2, 0]), color=BLUE)
        self.play(FadeIn(poi))
        self.wait(2)

        box = Rectangle(
            width=4, height=2.5,
            fill_color=BLACK, fill_opacity=1,
            stroke_color=WHITE, stroke_width=4
        )
        box.move_to(LEFT*4.5 + UP*2)  # or wherever you want it

        eqn = MathTex(
            r"\vec{E} = \frac{kQ}{", "r", r"^2}", r"\hat{r}",
            color=WHITE
        )

    
        eqn.scale_to_fit_width(box.width * 0.8)  # 80% of box width
        eqn.move_to(box.get_center())
        self.play(FadeIn(box), Write(eqn))
        self.wait(2)

        # After you create linecharge and poi
        x_tracker = ValueTracker(0)  # Start at center
        color_tracker = ValueTracker(0)  # 0 for green, 1 for red


        def get_arrow():
            start = np.array([x_tracker.get_value(), -2, 0])
            end = poi.get_center()
            color = PURE_GREEN if color_tracker.get_value() < 0.5 else PURE_RED
            return Arrow(start=start, end=end, color=color, buff=0, stroke_width=1.5)

        r_arrow = always_redraw(get_arrow)
        r_lab = always_redraw(lambda: MathTex("r").set_color(
            PURE_GREEN if color_tracker.get_value() < 0.5 else PURE_RED
        ).next_to(r_arrow.get_center(), RIGHT*0.8))


        self.play(GrowArrow(r_arrow), Write(r_lab), eqn[1].animate.set_color(PURE_GREEN))
        self.wait(1)

        # Animate the starting point from left to right, then back
        self.play(x_tracker.animate.set_value(5), run_time=3)
        self.wait(1)
        self.play(x_tracker.animate.set_value(-5), run_time=3)
        self.wait(1)
        self.play(x_tracker.animate.set_value(0), run_time=3)
        self.wait(2)
        # Animate color change
        self.play(color_tracker.animate.set_value(1), eqn[1].animate.set_color(PURE_RED), run_time=2)
        self.wait(2)

        self.play(FadeOut(r_arrow, r_lab, box, eqn))
        self.wait(1)



# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING s2p1
```

## (s2p2)

```python
class s2p2(ZoomedScene):
    def __init__(self, **kwargs):
        ZoomedScene.__init__(
            self,
            zoom_factor=0.3,
            zoomed_display_height=3,
            zoomed_display_width=3,
            image_frame_stroke_width=20,
            zoomed_camera_config={
                "default_frame_stroke_width": 3,
            },
            **kwargs
        )
    def construct(self):

        linecharge = Line(start=LEFT*5, end=RIGHT*5, color=RED, stroke_width=8).shift(DOWN*2)
        charge_lab = MathTex("Q").next_to(linecharge, DOWN)
        poi = Dot(point=np.array([0, 2, 0]), color=BLUE)
       
        self.add(poi, linecharge, charge_lab)
        self.wait(2)

        sample = Dot(radius = 0.04, point=linecharge.get_center()-np.array([-2, 0, 0]), color=YELLOW)
        sample_lab = MathTex("dq").next_to(sample.get_center(), DOWN*0.01).scale(0.35)
        self.play(FadeIn(sample))

        self.zoomed_camera.frame.move_to(sample.get_center())
        self.play(Animation(self.activate_zooming(animate=True)))
        self.wait(2)
        self.play(Write(sample_lab))
        self.wait(2)

        dashed = DashedLine(start=sample.get_center(), end=poi.get_center(), color=WHITE)
        dashed_lab = MathTex("r").set_color(WHITE).next_to(dashed.get_center(), RIGHT*0.8)
        self.play(Create(dashed), Write(dashed_lab))
        self.wait(2)

        start = poi.get_center()
        end = sample.get_center()
        direction = end - start
        direction = direction / np.linalg.norm(direction)  # Normalize
        arrow_length = 1  # Adjust as desired

        dE = Arrow(
            start=start,
            end=start - direction * arrow_length,
            color=YELLOW,
            buff=0,
            stroke_width=2
        )
        E_lab = MathTex(r"\vec{E}").set_color(YELLOW).next_to(dE.get_center(), LEFT*1.2+DOWN*0.5)
        E_lab_2 = MathTex(r"d\vec{E}").set_color(YELLOW).next_to(dE.get_center(), LEFT*1.2+DOWN*0.5)
        self.play(GrowArrow(dE), Write(E_lab))
        self.wait(2)


        self.play(Indicate(sample, scale_factor=1.5), Indicate(sample_lab, scale_factor=1.5))
        self.wait(1)
        self.play(FadeIn(E_lab_2))
        self.play(FadeOut(E_lab))
        self.wait(2)

        arrow_length = 0.3  # Adjust as desired

        dE_small = Arrow(
            start=start,
            end=start - direction * arrow_length,
            color=YELLOW,
            buff=0,
            stroke_width=2
        )
        E_lab_2.add_updater(lambda m: m.move_to(dE.get_end()+UP*0.15))

        self.play(Transform(dE, dE_small),E_lab_2.animate.scale(0.35))
        self.wait(2)
        self.play(self.zoomed_camera.frame.animate.move_to(poi.get_center()+UP*0.15))
        self.wait(2)



        def add_sample_and_field(scene, x_pos, line_y, poi, color=YELLOW, dash_color=WHITE, arrow_color=YELLOW):
            # Create sample dot
            sample = Dot(point=np.array([x_pos, line_y, 0]), color=color, radius=0.04)
            sample_lab = MathTex("dq").next_to(sample.get_center(), DOWN*0.01).scale(0.35)
            scene.play(FadeIn(sample), Write(sample_lab))
            scene.wait(0.5)

            # Dashed line from sample to poi
            dashed = DashedLine(start=sample.get_center(), end=poi.get_center(), color=dash_color)
            scene.play(Create(dashed))
            scene.wait(0.5)

            # Small dE arrow from poi toward sample
            direction = sample.get_center() - poi.get_center()
            direction = direction / np.linalg.norm(direction)
            arrow_length = 0.3
            dE_arrow = Arrow(
                start=poi.get_center(),
                end=poi.get_center() - direction * arrow_length,
                color=arrow_color,
                buff=0,
                stroke_width=2
            )
            scene.play(GrowArrow(dE_arrow))
            scene.wait(0.3)
            return dE_arrow

        # Example usage at the end of your scene:
        add_sample_and_field(
            self,
            x_pos=-3,         # x position for new sample dot
            line_y=-2,       # y position for linecharge
            poi=poi,         # your point of interest
            color=YELLOW,    # sample dot color
            dash_color=WHITE,# dashed line color
            arrow_color=YELLOW # dE arrow color
)
        self.wait(2)

        add_sample_and_field(
            self,
            x_pos=-1,         # x position for new sample dot
            line_y=-2,       # y position for linecharge
            poi=poi,         # your point of interest
            color=YELLOW,    # sample dot color
            dash_color=WHITE,# dashed line color
            arrow_color=YELLOW # dE arrow color
)
        self.wait(2)
        self.play(FadeOut(dashed_lab))

        dE_arrows = []
        for i in range(-5, 6, 1):
            dE_arrow = add_sample_and_field(
                self,
                x_pos=i,         # x position for new sample dot
                line_y=-2,       # y position for linecharge
                poi=poi,         # your point of interest
                color=YELLOW,    # sample dot color
                dash_color=WHITE,# dashed line color
                arrow_color=YELLOW # dE arrow color
    )            
            dE_arrows.append(dE_arrow)
        self.wait(2)


        sup_txt = Text("Superposition!", color=YELLOW).scale(1.2).shift(LEFT*3.5+UP*1).rotate(35*DEGREES)
        self.play(Write(sup_txt))
        self.wait(2)
        # Blink each dE_arrow in green, in succession
        for arrow in dE_arrows:
            self.play(Indicate(arrow, scale_factor=1.5, color=PURE_GREEN, run_time=0.25))
        self.wait(1)


# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING s2p2
```

## (s2p3)

```python
class s2p3(ZoomedScene):
    def __init__(self, **kwargs):
        ZoomedScene.__init__(
            self,
            zoom_factor=0.3,
            zoomed_display_height=3,
            zoomed_display_width=3,
            image_frame_stroke_width=20,
            zoomed_camera_config={
                "default_frame_stroke_width": 3,
            },
            **kwargs
        )
    def construct(self):

        linecharge = Line(start=LEFT*5, end=RIGHT*5, color=RED, stroke_width=8).shift(DOWN*2)
        charge_lab = MathTex("Q").next_to(linecharge, DOWN)
        poi = Dot(point=np.array([0, 2, 0]), color=BLUE)
       
        self.add(poi, linecharge, charge_lab)
        self.wait(2)
        self.zoomed_camera.frame.move_to(poi.get_center())
        self.play(Animation(self.activate_zooming(animate=True)))

        def add_sample_and_field(scene, x_pos, line_y, poi, color=YELLOW, dash_color=WHITE, arrow_color=YELLOW):
            # Create sample dot and label
            sample = Dot(point=np.array([x_pos, line_y, 0]), color=color, radius=0.04)
            sample_lab = MathTex("dq").next_to(sample.get_center(), DOWN*0.01).scale(0.35)
            scene.play(FadeIn(sample), Write(sample_lab))
            scene.wait(0.2)

            # Always redraw dashed line and dE arrow based on poi and sample
            dashed = always_redraw(lambda: DashedLine(
                start=sample.get_center(),
                end=poi.get_center(),
                color=dash_color
            ))
            dE_arrow = always_redraw(lambda: Arrow(
                start=poi.get_center(),
                end=poi.get_center() - (sample.get_center() - poi.get_center()) / np.linalg.norm(sample.get_center() - poi.get_center()) * 0.3,
                color=arrow_color,
                buff=0,
                stroke_width=2
            ))

            scene.play(Create(dashed), GrowArrow(dE_arrow))
            scene.wait(0.2)
            return dashed, dE_arrow, sample, sample_lab

        # Example usage:
        dashed_list = []
        dE_arrows = []
        for i in range(-5, 6, 1):
            dashed, dE_arrow, sample, sample_lab = add_sample_and_field(
                self,
                x_pos=i,
                line_y=-2,
                poi=poi,
                color=YELLOW,
                dash_color=WHITE,
                arrow_color=YELLOW
            )
            dashed_list.append(dashed)
            dE_arrows.append(dE_arrow)
        self.wait(1)

        box = Rectangle(
            width=5, height=2,
            fill_color=BLACK, fill_opacity=1,
            stroke_color=WHITE, stroke_width=4
        )
        box.move_to(LEFT*4 + UP*2)  # or wherever you want it

        # Create your MathTex
        eqn = MathTex(r"\vec{E}_{net}", "=", r"\vec{E}_1", "+", r"\vec{E}_2", "+", r"\vec{E}_3", "+", "...", "+", r"\vec{E}_n", color=WHITE)
        eqn_2 = MathTex(r"\vec{E}_{net}", "=", r"\sum_{i=1}^{n}", r"\vec{E}_i", color=WHITE)
        eqn.scale_to_fit_width(box.width * 0.8)  # 80% of box width
        eqn.move_to(box.get_center())
        eqn_2.scale_to_fit_width(box.width * 0.7)  # 70% of box width
        eqn_2.move_to(box.get_center())

        self.play(FadeIn(box), Write(eqn))
        self.wait(2)

        self.play(ReplacementTransform(eqn, eqn_2))
        self.wait(2)

        # Get endpoints of your linecharge
        x_start = linecharge.get_start()[0]
        x_end = linecharge.get_end()[0]
        y_val = linecharge.get_start()[1]

        num_dots = 120  # Adjust for density
        dot_radius = 0.04

        dots = VGroup()
        for i in range(num_dots + 1):
            x = x_start + (x_end - x_start) * i / num_dots
            dot = Dot(point=np.array([x, y_val, 0]), color=YELLOW, radius=dot_radius)
            dots.add(dot)

        for dot in dots:
            self.play(FadeIn(dot), run_time=0.07)
        self.wait(1)

        eq_3 = MathTex(r"\vec{E}_{net}", "=", r"\int", r"d\vec{E}", color=WHITE)
        eq_3.scale_to_fit_width(box.width * 0.7)  # 70% of box width
        eq_3.move_to(box.get_center())  
        self.play(ReplacementTransform(eqn_2, eq_3))
        self.wait(2)


        # The starting position of poi is (0, 2, 0), but is 4 units above the linecharge
        self.zoomed_camera.frame.add_updater(lambda m: m.move_to(poi.get_center()))
        self.play(poi.animate.move_to([0, -2+0.1, 0]), run_time=3)
        self.wait(1)
        self.play(poi.animate.move_to([-5, -2+1, 0]), run_time=3)
        self.wait(1)
        self.play(poi.animate.move_to([-6, -2+0, 0]), run_time=6)
        self.wait(2)
        self.play(poi.animate.move_to([0, 40, 0]), run_time=4)
        self.wait(2)
        self.play(poi.animate.move_to([0, 2, 0]), run_time=4)
        self.wait(2)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING s2p3
```

```python
class s2p4(ZoomedScene):
    def __init__(self, **kwargs):
        ZoomedScene.__init__(
            self,
            zoom_factor=0.3,
            zoomed_display_height=3,
            zoomed_display_width=3,
            image_frame_stroke_width=20,
            zoomed_camera_config={
                "default_frame_stroke_width": 3,
            },
            **kwargs
        )
    def construct(self):

        linecharge = Line(start=LEFT*5, end=RIGHT*5, color=RED, stroke_width=8).shift(DOWN*2)
        charge_lab = MathTex("Q").next_to(linecharge, DOWN)
        poi = Dot(point=np.array([0, 2, 0]), color=BLUE)
       
        self.add(poi, linecharge, charge_lab)
        self.zoomed_camera.frame.move_to(poi.get_center())
        self.play(Animation(self.activate_zooming(animate=False)))

        def add_sample_and_field(scene, x_pos, line_y, poi, color=YELLOW, dash_color=WHITE, arrow_color=YELLOW):
            # Create sample dot and label
            sample = Dot(point=np.array([x_pos, line_y, 0]), color=color, radius=0.04)
            sample_lab = MathTex("dq").next_to(sample.get_center(), DOWN*0.01).scale(0.35)
            scene.add(sample)

            # Always redraw dashed line and dE arrow based on poi and sample
            dashed = always_redraw(lambda: DashedLine(
                start=sample.get_center(),
                end=poi.get_center(),
                color=dash_color
            ))
            dE_arrow = always_redraw(lambda: Arrow(
                start=poi.get_center(),
                end=poi.get_center() - (sample.get_center() - poi.get_center()) / np.linalg.norm(sample.get_center() - poi.get_center()) * 0.3,
                color=arrow_color,
                buff=0,
                stroke_width=2
            ))

            scene.add(dashed ,dE_arrow)
            return dashed, dE_arrow, sample, sample_lab

        dashed_list = []
        dE_arrows = []
        for i in range(-5, 6, 1):
            dashed, dE_arrow, sample, sample_lab = add_sample_and_field(
                self,
                x_pos=i,
                line_y=-2,
                poi=poi,
                color=YELLOW,
                dash_color=WHITE,
                arrow_color=YELLOW
            )
            dashed_list.append(dashed)
            dE_arrows.append(dE_arrow)
        self.wait(1)


        # Get endpoints of your linecharge
        x_start = linecharge.get_start()[0]
        x_end = linecharge.get_end()[0]
        y_val = linecharge.get_start()[1]

        num_dots = 120  # Adjust for density
        dot_radius = 0.04

        dots = VGroup()
        for i in range(num_dots + 1):
            x = x_start + (x_end - x_start) * i / num_dots
            dot = Dot(point=np.array([x, y_val, 0]), color=YELLOW, radius=dot_radius)
            dots.add(dot)


        self.add(dots)
        self.wait(1)

        box = Rectangle(
            width=5, height=2,
            fill_color=BLACK, fill_opacity=1,
            stroke_color=WHITE, stroke_width=4
        )
        box.move_to(LEFT*4 + UP*2)  # or wherever you want it

        marker = Dot(point=box.get_center(), color=BLUE, radius=0.06)
        arr_1 = Arrow(start=marker.get_center(), end=marker.get_center()+UP*1+RIGHT*1, color = YELLOW, buff=0, stroke_width=2).shift(DOWN*0.5)
        arr_2 = Arrow(start=marker.get_center(), end=marker.get_center()+UP*1+LEFT*1, color = YELLOW, buff=0, stroke_width=2).shift(DOWN*0.5)


        self.play(FadeIn(box), GrowArrow(arr_1), GrowArrow(arr_2))
        self.wait(2)

        arr_1_x = Arrow(start=marker.get_center(), end=marker.get_center()+RIGHT*1, color = RED, buff=0, stroke_width=2).shift(DOWN*0.5)
        arr_2_x = Arrow(start=marker.get_center(), end=marker.get_center()+LEFT*1, color = RED, buff=0, stroke_width=2).shift(DOWN*0.5)

        arr_1_y = Arrow(start=marker.get_center()+RIGHT, end=marker.get_center()+UP*1+RIGHT, color = PURE_GREEN, buff=0, stroke_width=2).shift(DOWN*0.5)
        arr_2_y = Arrow(start=marker.get_center()+LEFT, end=marker.get_center()+UP*1+LEFT, color = PURE_GREEN, buff=0, stroke_width=2).shift(DOWN*0.5)

        self.play(GrowArrow(arr_1_x), GrowArrow(arr_2_x))
        self.wait(2)
        self.play(GrowArrow(arr_1_y), GrowArrow(arr_2_y))
        self.wait(2)

        self.play(FadeOut(arr_1_x, arr_2_x))
        self.wait(1)
        self.play(FadeOut(arr_1, arr_2))
        self.wait(1)
        self.play(arr_1_y.animate.move_to(marker.get_center()), arr_2_y.animate.move_to(marker.get_center()))
        self.wait(2)

        total_field = Arrow(
            start=poi.get_center(),
            end=poi.get_center()+UP*0.6,
            color=PURE_GREEN,
            buff=0,
            stroke_width=3
        )
        self.play(FadeOut(*dE_arrows), FadeIn(total_field), self.zoomed_camera.frame.animate.scale(1.8), FadeOut(*dashed_list))
        self.wait(2)

        eq = MathTex(r"\vec{E}_{net}", "=", r"\int d\vec{E}", color=WHITE)
        eq.scale_to_fit_width(box.width * 0.7)  # 70% of box width
        eq.move_to(box.get_center())
        self.play(FadeOut(arr_1_y, arr_2_y), Write(eq))
        self.wait(2)

        eq_2 = MathTex(r"\vec{E}_{net}", "=", r"\int \frac{k dq}{r^2} \ \hat{r}", color=WHITE)
        eq_2.scale_to_fit_width(box.width * 0.9)  # 90% of box width
        eq_2.move_to(box.get_center())
        self.play(ReplacementTransform(eq, eq_2))
        self.wait(2)

        eq_3 = MathTex(r"\vec{E}_{line}", "=", r"\frac{k \ Q_{tot}}{x\sqrt{x^2 + (L/2)^2}} \ \hat{y}", color=WHITE)
        eq_3.scale_to_fit_width(box.width * 0.9)  # 90% of box width
        eq_3.move_to(box.get_center())
        self.play(ReplacementTransform(eq_2, eq_3))
        self.wait(2)

        dashed_final = DashedLine(start=linecharge.get_center(), end=poi.get_center(), color=WHITE, stroke_width=2)
        dashed_label = MathTex("x").set_color(WHITE).next_to(dashed_final.get_center(), RIGHT*0.8)
        self.play(Create(dashed_final), Write(dashed_label))
        self.wait(2)    
        length_r = Arrow(
            start=linecharge.get_center()+RIGHT*0.5,
            end=linecharge.get_right(),
            color=WHITE,
            buff=0,
            stroke_width=3
        ).shift(UP*0.2)

        length_l = Arrow(
            start=linecharge.get_center()+LEFT*0.5,
            end=linecharge.get_left(),
            color=WHITE,
            buff=0,
            stroke_width=3
        ).shift(UP*0.2)

        l_label = MathTex("L").set_color(WHITE).next_to(linecharge.get_center(), UP*0.5)

        self.play(GrowArrow(length_r), GrowArrow(length_l), Write(l_label))
        self.wait(2)



# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING s2p4
```

## (s2p5)

```python
class s2p5(ThreeDScene):
    def construct(self):
        self.set_camera_orientation(phi=75 * DEGREES, theta=-45 * DEGREES)

        # Create the slab (Prism)
        slab = Prism(
            dimensions=[4, 4, 0.1],
            fill_color=RED,
            fill_opacity=1,
            stroke_color=WHITE,
            stroke_width=2
        ).move_to(np.array([0, 0, -2]))

        # Create a point of interest (POI)
        poi = Sphere(radius=0.05, color=BLUE).move_to(np.array([0, 0, 2]))

        self.add(slab, poi)
        self.wait(2)

        # Create a dashed line and label
        dashed = DashedLine(start=slab.get_center(), end=poi.get_center(), color=WHITE)
        dashed_label = MathTex("z").set_color(WHITE).next_to(dashed.get_center(), LEFT * 1.2).scale(2)
        dashed_label.rotate(90 * DEGREES, axis=RIGHT)
        dashed_label.rotate(45 * DEGREES, axis=OUT)
        self.play(Create(dashed), Write(dashed_label))
        self.wait(2)

        # Define the grid parameters
        slab_width, slab_height, slab_depth = 4, 4, 0.1  # Dimensions of the slab
        grid_size = 20  # Number of dots along one side
        top_face_center = slab.get_center() + np.array([0, 0, slab_depth / 2])  # Top face center
        dot_spacing = slab_width / (grid_size - 1)  # Spacing between dots

        # Create the grid of dots
        dots = VGroup()
        for i in range(grid_size):
            for j in range(grid_size):
                x_offset = (i - (grid_size - 1) / 2) * dot_spacing
                y_offset = (j - (grid_size - 1) / 2) * dot_spacing
                dot = Dot(point=top_face_center + np.array([x_offset, y_offset, 0]), color=YELLOW)
                dots.add(dot)

        # Fade in the grid of dots
        self.play(FadeIn(dots))
        self.wait(2)

        box1 = Rectangle(
            width=5, height=2,
            fill_color=BLACK, fill_opacity=1,
            stroke_color=WHITE, stroke_width=4
        )

        box1.move_to(LEFT*4 + UP*2)  # or wherever you want it
        point_eq = MathTex(r"\vec{E} = \int \frac{k \ dq}{r^2} \ \hat{r}", color=WHITE)
        point_eq.scale_to_fit_width(box1.width * 0.9)  # 90% of box width
        point_eq.move_to(box1.get_center())
        box_eq_1 = VGroup(box1, point_eq)

        self.add_fixed_in_frame_mobjects(box_eq_1)
        self.play(FadeIn(box_eq_1))
        self.wait(2)

        box2 = Rectangle(
            width=5, height=2,
            fill_color=BLACK, fill_opacity=1,
            stroke_color=WHITE, stroke_width=4
        )



        line_spacing = slab_width / (grid_size - 1)  # Spacing between lines

        # Create the grid of lines
        lines = VGroup()
        for i in range(grid_size):
            y_offset = (i - (grid_size - 1) / 2) * line_spacing
            start = top_face_center + np.array([-slab_width / 2, y_offset, 0])
            end = top_face_center + np.array([slab_width / 2, y_offset, 0])
            line = Line(start=start, end=end, color=YELLOW)
            lines.add(line)

        # Fade in the grid of lines
        self.play(FadeOut(dots),FadeIn(lines))
        self.play(Indicate(lines, scale_factor=1.2, color=PURE_GREEN, run_time=3))
        self.wait(2)


# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING s2p5
        
```

```python
class s3p1(ZoomedScene):
    def __init__(self, **kwargs):
        ZoomedScene.__init__(
            self,
            zoom_factor=0.3,
            zoomed_display_height=3,
            zoomed_display_width=3,
            image_frame_stroke_width=20,
            zoomed_camera_config={
                "default_frame_stroke_width": 3,
            },
            **kwargs
        )
    def construct(self):

        linecharge = Line(start=LEFT*5, end=RIGHT*5, color=RED, stroke_width=8).shift(DOWN*2)
        charge_lab = MathTex("Q").next_to(linecharge, DOWN)
        poi = Dot(point=np.array([0, 2, 0]), color=BLUE)
       
        self.add(poi, linecharge, charge_lab)
        self.wait(2)
        self.play(FadeOut(poi))
        #self.zoomed_camera.frame.move_to(poi.get_center())
        #self.play(Animation(self.activate_zooming(animate=True)))


        box = Rectangle(
            width=5, height=2,
            fill_color=BLACK, fill_opacity=1,
            stroke_color=WHITE, stroke_width=4
        )
        box.move_to(LEFT*4 + UP*2)  # or wherever you want it

        # Create your MathTex
        eqn = MathTex(r"Q_{tot} = \int dq", color=WHITE)
        eqn.scale_to_fit_width(box.width * 0.65)
        eqn.move_to(box.get_center())


        self.play(FadeIn(box), Write(eqn))
        self.wait(2)

        ax = Axes(
            x_range=[0, 6, 1],
            y_range=[0, 10, 1],
            x_length=6,
            y_length=4,
            axis_config={"include_numbers": False},
        ).move_to(RIGHT*4 + UP*2).scale(0.5)
        labels = ax.get_axis_labels(
            Tex("x(m)").scale(0.7), Text("Q(C)").scale(0.45)
        )
        m = 1.3  # Slope of the line
        graph = ax.plot(lambda x: m*x, x_range=[0, 5], color=YELLOW)
        length_mark = DashedLine(
            start=ax.c2p(5, 0),
            end=ax.c2p(5, m*5),
            color=WHITE
        )

        self.play(Create(ax), Write(labels))
        self.wait(1)

        # Get endpoints of your linecharge
        x_start = linecharge.get_start()[0]
        x_end = linecharge.get_end()[0]
        y_val = linecharge.get_start()[1]

        # Axes x-range
        ax_x_start = ax.x_range[0]
        ax_x_end = ax.x_range[1]
        graph_x_start = 0
        graph_x_end = 5

        num_dots = 120
        dot_radius = 0.04

        dots = VGroup()
        graph_segments = VGroup()
        x_vals = np.linspace(x_start, x_end, num_dots + 1)
        ax_x_vals = np.linspace(graph_x_start, graph_x_end, num_dots + 1)


        for i in range(num_dots + 1):
            # Dot in scene coordinates
            x = x_vals[i]
            dot = Dot(point=np.array([x, y_val, 0]), color=YELLOW, radius=dot_radius)
            dots.add(dot)
            # Graph segment in axes coordinates
            if i > 0:
                x_prev = ax_x_vals[i-1]
                x_curr = ax_x_vals[i]
                y_prev = m * x_prev
                y_curr = m * x_curr
                seg = Line(
                    ax.c2p(x_prev, y_prev),
                    ax.c2p(x_curr, y_curr),
                    color=YELLOW
                )
                graph_segments.add(seg)

        # Animate both together
        for i, dot in enumerate(dots):
            anims = [FadeIn(dot, run_time=0.07)]
            if i > 0:
                anims.append(Create(graph_segments[i-1], run_time=0.07))
            self.play(*anims)
        self.wait(1)

        self.play(Create(length_mark))
        self.wait(2)
        length_lab = MathTex("L").set_color(WHITE).next_to(length_mark.get_bottom(), DOWN*1)
        self.play(Write(length_lab))
        self.wait(2)

        start_dot = Dot(point=linecharge.get_start(), color=PURE_GREEN, radius=.1)
        end_dot = Dot(point=linecharge.get_end(), color=PURE_GREEN, radius=.1)
        zero_lab = MathTex("0").set_color(WHITE).next_to(linecharge.get_start(), UP*1)
        self.play(FadeIn(start_dot), Write(zero_lab))
        self.play(FadeIn(end_dot), Write(MathTex("L").set_color(WHITE).next_to(linecharge.get_end(), UP*1)))
        self.wait(2)

        q_line = DashedLine(
            start=ax.c2p(0, m*5),
            end=ax.c2p(5, m*5),
            color = WHITE
        )
        q_lab = MathTex("Q").set_color(WHITE).next_to(q_line.get_start(), LEFT*0.5)
        self.play(Create(q_line), Write(q_lab))
        self.play(Indicate(q_lab, scale_factor=1.5), Indicate(charge_lab, scale_factor=1.5))
        self.wait(2)

        def add_vertical_ticks_along_line(scene, line, spacing=1, height=0.2, color=WHITE, z_index=1):
            """
            Adds small vertical lines (ticks) every `spacing` units along the given line.
            """
            start = line.get_start()
            end = line.get_end()
            line_vec = end - start
            line_len = np.linalg.norm(line_vec)
            line_dir = line_vec / line_len
            num_ticks = int(line_len // spacing)
            ticks = VGroup()
            for i in range(1, num_ticks):
                pos = start + line_dir * (i * spacing)
                # Perpendicular direction (vertical in scene)
                perp = np.array([0, 1, 0])
                tick = Line(
                    pos - perp * (height / 2),
                    pos + perp * (height / 2),
                    color=color,
                    stroke_width=3,
                    z_index=z_index
                )
                ticks.add(tick)
            #scene.add(ticks)
            return ticks

        # Usage in your construct:
        # ...after creating linecharge...
        ticks = add_vertical_ticks_along_line(self, linecharge, spacing=1, height=0.2, color=WHITE)
        fadein_speed = 0.1  # seconds per tick

        for tick in ticks:
            self.play(FadeIn(tick), run_time=fadein_speed)
        self.wait(1)

        labels = VGroup()
        label_latex = r"3\ \frac{C}{m}"  # Your desired LaTeX

        # First label: to the left of the first tick
        first_tick = ticks[0]
        second_tick = ticks[1]
        left_pos = first_tick.get_center() - (second_tick.get_center() - first_tick.get_center())
        label_left = MathTex(label_latex, font_size=28, color=WHITE).move_to((first_tick.get_center() + left_pos) / 2 + UP*0.55)
        labels.add(label_left)
        self.play(FadeIn(label_left))
        #self.add(label_left)

        # Labels between ticks
        for i in range(len(ticks) - 1):
            p1 = ticks[i].get_center()
            p2 = ticks[i + 1].get_center()
            mid = (p1 + p2) / 2
            label = MathTex(label_latex, font_size=28, color=WHITE).move_to(mid + UP*0.55)
            labels.add(label)
            self.play(FadeIn(label))

        # Last label: to the right of the last tick
        last_tick = ticks[-1]
        second_last_tick = ticks[-2]
        right_pos = last_tick.get_center() + (last_tick.get_center() - second_last_tick.get_center())
        label_right = MathTex(label_latex, font_size=28, color=WHITE).move_to((last_tick.get_center() + right_pos) / 2 + UP*0.55)
        labels.add(label_right)
        self.play(FadeIn(label_right))

        self.wait(1)


# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING s3p1
```
