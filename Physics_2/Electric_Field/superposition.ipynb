## (s1p3)

```python
class s1p3(Scene):
    def construct(self):
        number_plane = NumberPlane(
            background_line_style={
                "stroke_width": 1,
                "stroke_opacity": 0.6
            }
        )# Axis labels
        x_label = MathTex("x").next_to(number_plane.c2p(6, 0), RIGHT+DOWN*0.5)
        y_label = MathTex("y").next_to(number_plane.c2p(0, 3), UP+RIGHT*0.7+UP*0.2)
        self.bring_to_back(number_plane)
        self.play(Create(number_plane), Write(x_label), Write(y_label))
        self.wait(1)


        # Define grid limits and spacing
        x_min, x_max, y_min, y_max = -6, 6, -5, 3
        spacing = 2

        # Create grid points
        points = [
            np.array([x, y, 0])
            for x in np.arange(x_min, x_max + spacing, spacing)
            for y in np.arange(y_min, y_max + spacing, spacing)
        ]
        def get_component_arrows_with_animations(arrow, color_x=YELLOW, color_y=ORANGE):
            """
            Given a Manim Arrow, return:
            - the x and y component arrows (as a tuple)
            - a list of ReplacementTransform animations from copies of the original arrow to the components
            """
            start = arrow.get_start()
            end = arrow.get_end()
            vec = end - start
            x_comp, y_comp = vec[0], vec[1]

            # Arrow for x component (horizontal)
            arrow_x = Arrow(
                start=start,
                end=start + np.array([x_comp, 0, 0]),
                color=color_x,
                buff=0
            )

            # Arrow for y component (vertical), starts at tip of x component
            arrow_y = Arrow(
                start=start + np.array([x_comp, 0, 0]),
                end=start + np.array([x_comp, y_comp, 0]),
                color=color_y,
                buff=0
            )

            # Animations: ReplacementTransform from copies of the original arrow
            anims = [
                ReplacementTransform(arrow.copy(), arrow_x),
                ReplacementTransform(arrow.copy(), arrow_y),
                arrow.animate.set_opacity(0.2)
            ]

            return (arrow_x, arrow_y), anims

        arrows = []
        all_anims = []
        for pt in points:
            for i in range(3):
                theta = random.uniform(0, 2 * np.pi)
                length = random.uniform(0.5, 1)
                direction = np.array([np.cos(theta), np.sin(theta), 0])
                color = [BLUE, GREEN, PURE_RED][i]  # cycle colors for clarity
                arrow = Arrow(
                    start=pt,
                    end=pt + length * direction,
                    buff=0,
                    color=color
                )
                arrows.append(arrow)
                (arrow_x, arrow_y), anims = get_component_arrows_with_animations(arrow)
                all_anims.extend(anims)

        self.play(FadeIn(*arrows))
        self.wait(1)
       # self.play(*all_anims)
       # self.wait(2)

        def sum_and_transform_components(component_arrows, point, color=YELLOW):
            """
            Given a list of component arrows (all starting at 'point'), 
            create a resultant arrow and return Transform animations 
            to morph all components into the resultant.
            """
            # Sum all vectors
            vectors = [arrow.get_end() - arrow.get_start() for arrow in component_arrows]
            resultant_vec = sum(vectors, np.array([0, 0, 0]))
            resultant_arrow = Arrow(
                start=point,
                end=point + resultant_vec,
                color=color,
                buff=0
            )
            # Create transforms from each component to the resultant
            transforms = [Transform(arrow, resultant_arrow.copy()) for arrow in component_arrows]
            return resultant_arrow, transforms

        # --- Usage pattern ---

        # Suppose for each grid point you have:
        # - pt: the grid point
        # - comp_arrows: a list of 6 component arrows (3 x, 3 y) at that point

        resultant_arrows = []
        all_transforms = []

        # Group component arrows by grid point
        from collections import defaultdict
        components_by_point = defaultdict(list)
        for arrow in arrows:  # 'arrows' is your list of all component arrows
            pt = tuple(np.round(arrow.get_start(), 8))  # Use rounded tuple for hashable key
            components_by_point[pt].append(arrow)

        for pt_tuple, comp_arrows in components_by_point.items():
            pt = np.array(pt_tuple)
            resultant_arrow, transforms = sum_and_transform_components(comp_arrows, pt)
            resultant_arrows.append(resultant_arrow)
            all_transforms.extend(transforms)

        # Add resultant arrows to the scene (so Transform has a target)

        self.play(*all_transforms)
        self.wait(2)
        self.add(*resultant_arrows)
        self.wait(2)

        self.play(FadeOut(*arrows), FadeOut(*resultant_arrows))
        self.wait(2)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING s1p3
```

## (s1p5)

```python
class s1p5(Scene):
    def construct(self):
        import numpy as np

        q1_pos = np.array([-3, 0])
        q2_pos = np.array([3, 0])
        q1 = 1
        q2 = -1

        dot1 = Dot(point=np.array([-3, 0, 0]), color=RED, radius=0.15)
        dot2 = Dot(point=np.array([3, 0, 0]), color=BLUE, radius=0.15)
        #self.add(dot1, dot2)

        def E_field(pos):
            r1 = pos - q1_pos
            r2 = pos - q2_pos
            e1 = q1 * r1 / (np.linalg.norm(r1)**3 + 1e-6)
            e2 = q2 * r2 / (np.linalg.norm(r2)**3 + 1e-6)
            return e1 + e2

        def make_field_line(start, step=0.15, n_steps=120, reverse=False):
            points = [np.array([*start, 0])]
            pos = np.array(start)
            for _ in range(n_steps):
                E = E_field(pos)
                E_norm = np.linalg.norm(E)
                if E_norm < 1e-3:
                    break
                dpos = step * E / E_norm
                if reverse:
                    dpos = -dpos
                pos = pos + dpos
                points.append(np.array([*pos, 0]))
                if np.linalg.norm(pos - q1_pos) < 0.3 or np.linalg.norm(pos - q2_pos) < 0.3:
                    break
            return points

        n_lines = 16
        radius = 0.18


        field_lines = []
        arrow_tips = []

        # Field lines from positive charge (outward, along E) - full circle
        for i in range(n_lines):
            angle = 2 * np.pi * i / n_lines
            start = q1_pos + radius * np.array([np.cos(angle), np.sin(angle)])
            points = make_field_line(start)
            if len(points) < 2:
                continue
            field_line = VMobject().set_points_smoothly(points)
            field_line.set_stroke(YELLOW, width=2)
            field_lines.append(field_line)
            # Arrow tip at midpoint (only for red charge lines)
            mid_idx = len(points) // 2
            if mid_idx < len(points) - 1:
                p0 = points[mid_idx]
                p1 = points[mid_idx + 1]
                direction = p1 - p0
                angle_tip = np.arctan2(direction[1], direction[0])
                arrow_tip = Arrow(
                    start=p0,
                    end=p0 + 0.4 * direction / np.linalg.norm(direction),
                    buff=0,
                    max_stroke_width_to_length_ratio=10,
                    stroke_width=1,
                    fill_color=YELLOW,
                    tip_length=0.35,
                )
                arrow_tip.rotate(angle_tip - arrow_tip.get_angle(), about_point=p0)
                arrow_tip.set_color(YELLOW)
                arrow_tips.append(arrow_tip)

        # Field lines from negative charge (inward, against E) - only -60 to +60 degrees
        n_blue_lines = 6
        for i in range(n_blue_lines):
            theta = -np.pi/3 + (2 * np.pi / 3) * i / (n_blue_lines - 1)  # from -60° to +60°
            start = q2_pos + radius * np.array([np.cos(theta), np.sin(theta)])
            points = make_field_line(start, reverse=True)
            if len(points) < 2:
                continue
            field_line = VMobject().set_points_smoothly(points)
            field_line.set_stroke(YELLOW, width=2)
            field_lines.append(field_line)
            # No arrow tip for blue charge lines

        # Fade in all field lines and arrow tips together

        q1 = mp.Charge(magnitude=1, point=np.array([-3, 0, 0]))
        q2 = mp.Charge(magnitude=-1, point=np.array([3, 0, 0]))
        efield_3 = mp.ElectricField(q1, q2, color=YELLOW)
        self.add(q1, q2, efield_3)
        self.wait(1)

        self.play(*[FadeIn(mob, run_time=1) for mob in field_lines + arrow_tips], FadeOut(efield_3))
        self.wait(2)

        box = Rectangle(
            width=3, height=1,
            fill_color=BLACK, fill_opacity=1,
            stroke_color=WHITE, stroke_width=4
        )
        box.move_to(UP*3)  # or wherever you want it

        # Create your MathTex
        eqn = Text("Dipole Field ", color=WHITE)
        eqn.scale_to_fit_width(box.width * 0.8)  # 80% of box width
        eqn.move_to(box.get_center())

        self.play(FadeIn(box), Write(eqn))
        self.wait(2)

        self.play(FadeOut(q1, q2))
        self.wait(2)


        
# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING s1p5
```
