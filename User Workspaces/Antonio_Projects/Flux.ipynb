# Create a Electric Field of 2D arrows

Try creating an arrow of arrows with a loop //
Dont change the camera orientation //
gather the endpoints of the arrows in an array //
print out the corner endpoints //
use those printed endpoints to create a sheet over the array of arrows //
place the sheet and arrows into a VGroup //
rotate both the Vgroup //
remove the sheet from the group //
now move the sheet around //

```python
class eField(ThreeDScene):
    def construct(self):
        # Set camera orientation to view the plane above
        self.set_camera_orientation(phi=60 * DEGREES, theta=90 * DEGREES)

        # Create a grid of arrows
        rows = 6
        cols = 6
        spacing = 1

        arrow_grid = VGroup()
        for i in range(rows):
            for j in range(cols):
                # Compute position of the arrow's start and end
                x = j * spacing
                y = i * spacing
                start = np.array([x, y, 0])
                end = np.array([x, y + 2, 0])

                arrow = Arrow(start, end, color=BLUE, buff=0.0, stroke_width=4)
                arrow.rotate(PI / 3, axis=RIGHT)
                arrow_grid.add(arrow)
                
        # Center the whole grid and lift it into 3D space
        arrow_grid.move_to(ORIGIN)

        # Add the grid to the scene
        self.add(arrow_grid)

        # Place a box on top of the arrows
        box = Prism(dimensions=[10, 0.1, 10], fill_opacity=0.25, stroke_width=2).scale(0.5)
        box.set_fill(WHITE, opacity=0.25)
        box.set_stroke(PURE_RED, width=2)
        box.move_to(ORIGIN)  # Lift it
        box.rotate(PI / 3, axis=RIGHT)

        self.add(box)

        # Begin ambient rotation to show off the scene
        self.begin_ambient_camera_rotation(rate=0.1)
        self.wait(5)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING eField
```

# Scene 2 Part 2
Side on sheet in an electric field

```python
class flux_s2p2(Scene):
    def construct(self):
        arrows = VGroup()  # Group to hold all arrows

        # Create 6 arrows spaced horizontally
        for i in range(6):
            arrow = Arrow(start=DOWN, end=UP*3, buff=0, color=BLUE)
            arrow.move_to(RIGHT * (i - 2.5) * 1.5)  # Center them nicely
            arrows.add(arrow)


        # Create the red line from the first to last arrow
        left_arrow = arrows[0]
        right_arrow = arrows[-1]
        red_line = Line(
            start=left_arrow.get_center(), 
            end=right_arrow.get_center(), 
            color=RED
        )

        # Place an arrow on the center of the red line
        center_arrow = Arrow(start=red_line.get_center(), end=UP*1.5, buff=0, color=YELLOW)
        # center_arrow = LabeledArrow("A", start=red_line.get_center(), end=UP, buff=0, color=YELLOW)

        # Add everything to the scene
        self.play(
            AnimationGroup(
                FadeIn(arrows),
                FadeIn(red_line),
                FadeIn(center_arrow),
                lag_ratio=0.0
            )
        )
        self.wait()

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING flux_s2p2
```

```python
class flux_s2p3(Scene):
    def construct(self):
        arrows = VGroup()  # Group to hold all arrows

        # Create 6 arrows spaced horizontally
        for i in range(6):
            arrow = Arrow(start=DOWN, end=UP*3, buff=0, color=BLUE)
            arrow.move_to(RIGHT * (i - 2.5) * 1.5)  # Center them nicely
            arrows.add(arrow)


        # Create the red line from the first to last arrow
        left_arrow = arrows[0]
        right_arrow = arrows[-1]
        red_line = Line(
            start=left_arrow.get_center(), 
            end=right_arrow.get_center(), 
            color=RED
        )

        # Place an arrow on the center of the red line
        center_arrow = Arrow(start=red_line.get_center(), end=UP *1.5, buff=0, color=YELLOW)
        # center_arrow = LabeledArrow("A", start=red_line.get_center(), end=UP*1.5, buff=0, color=YELLOW)

        # Group the center arrow with the red line
        center_arrow_group = VGroup(red_line, center_arrow)

        # Add everything to the scene
        self.add(arrows)
        self.add(center_arrow_group)
        self.wait()

        # Replace the 4th arrow with a taller one
        base_point = arrows[3].get_start()
        new_arrow = Arrow(start=base_point, end=base_point + UP * 6, buff=0, color=BLUE)

        self.play(Transform(arrows[3], new_arrow))

        # Create a dashed line attached to the center arrow
        dashed_line = DashedLine(
            start=center_arrow.get_start() - UP * 1.25, 
            end=center_arrow.get_start() + UP * 5, 
            color=WHITE, 
            dash_length=0.1
        )
         # Update the center arrow group to include the dashed line
        center_arrow_group.add(dashed_line)

        # Slightly roate the center arrow group
        self.play(center_arrow_group.animate.rotate(-20 * DEGREES, axis=OUT, about_point=red_line.get_center()))

        self.wait(2)

        

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING flux_s2p3
```

# Scene 2 Part 4
Make the sheet parallel to the e-field

```python
class flux_s2p4(Scene):
    def construct(self):
        arrows = VGroup()  # Group to hold all arrows

        # Create 6 arrows spaced horizontally
        for i in range(6):
            arrow = Arrow(start=DOWN, end=UP*3, buff=0, color=BLUE)
            arrow.move_to(RIGHT * (i - 2.5) * 1.5)  # Center them nicely
            arrows.add(arrow)


        # Create the red line from the first to last arrow
        left_arrow = arrows[0]
        right_arrow = arrows[-1]
        red_line = Line(
            start=left_arrow.get_center(), 
            end=right_arrow.get_center(), 
            color=RED
        )

        # Place an arrow on the center of the red line
        center_arrow = Arrow(start=red_line.get_center(), end=UP*1.5, buff=0, color=YELLOW)

        # Replace the 4th arrow with a taller one
        base_point = arrows[3].get_start()
        arrows[3].put_start_and_end_on(base_point, base_point + UP * 6)

        # Group the center arrow with the red line
        center_arrow_group = VGroup(red_line, center_arrow)

        # Create a dashed line attached to the center arrow
        dashed_line = DashedLine(
            start=center_arrow.get_start() - UP * 1.25, 
            end=center_arrow.get_start() + UP * 5, 
            color=WHITE, 
            dash_length=0.1
        )
         # Update the center arrow group to include the dashed line
        center_arrow_group.add(dashed_line)

        # Slightly roate the center arrow group
        center_arrow_group.rotate(-20 * DEGREES, axis=OUT, about_point=red_line.get_center())


        # Add everything to the scene
        self.add(arrows)
        self.add(center_arrow_group)
        self.wait()
        self.wait(2)

        # Remove the dashed line
        center_arrow_group.remove(dashed_line)
        self.play(FadeOut(dashed_line))
        self.wait(1)

        # Return arrow 3 to its original height
        original_arrow = Arrow(start=base_point, end=base_point + UP * 4, buff=0, color=BLUE)
        self.play(Transform(arrows[3], original_arrow))

        # Keep rotating the center arrow group
        self.play(center_arrow_group.animate.rotate(-70 * DEGREES, axis=OUT, about_point=red_line.get_center()))

        self.wait(2)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING flux_s2p4
```

# Scene 3 Part 2
Create a Cosine function graph

```python
class flux_s3p2(Scene):
    def construct(self):
        # Set up axes
        axes = Axes(
            x_range=[0, 2 * PI, PI / 2],
            y_range=[-1.5, 1.5, 0.5],
            x_length=7,
            y_length=4,
            axis_config={"color": WHITE},
            tips=False
        )

        # Axis labels
        x_label = MathTex(r"\theta").next_to(axes.x_axis, RIGHT)
        y_label = MathTex(r"\Phi").next_to(axes.y_axis, UP)
        axes_labels = VGroup(x_label, y_label)

        # Points and x-axis labels at key angles
        theta_values = [0, PI/2, PI, 3*PI/2, 2*PI]
        degree_labels = [
            r"\text{0}^{\circ}", 
            r"\text{90}^{\circ}", 
            r"\text{180}^{\circ}", 
            r"\text{270}^{\circ}", 
            r"\text{360}^{\circ}"
        ]
        dots = VGroup()
        tick_labels = VGroup()

        for theta, deg_label in zip(theta_values, degree_labels):
            phi = np.cos(theta)
            # Dot on the curve
            dot = Dot(point=axes.c2p(theta, phi), color=RED)
            dots.add(dot)

            # Label directly below the x-axis tick
            label = MathTex(deg_label).scale(0.5)
            label.next_to(axes.c2p(theta, 0), DOWN, buff=0.25)
            tick_labels.add(label)

        # Show everything
        self.play(Create(axes), Write(axes_labels))
        self.play(FadeIn(dots), FadeIn(tick_labels))
        self.wait(2)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING flux_s3p2
```

# Scene 3 Part 4
Attach the Tracing Dot along the Cosine Function

```python
class flux_s3p4(Scene):
    def construct(self):
        # Set up axes
        axes = Axes(
            x_range=[0, 2 * PI, PI / 2],
            y_range=[-1.5, 1.5, 0.5],
            x_length=7,
            y_length=4,
            axis_config={"color": WHITE},
            tips=False
        )

        # Axis labels
        x_label = MathTex(r"\theta").next_to(axes.x_axis, RIGHT)
        y_label = MathTex(r"\Phi").next_to(axes.y_axis, UP)
        axes_labels = VGroup(x_label, y_label)

        # Plot the cosine curve: Phi(θ) = cos(θ)
        graph = axes.plot(lambda x: np.cos(x), color=BLUE)

        # Points and x-axis labels at key angles
        theta_values = [0, PI/2, PI, 3*PI/2, 2*PI]
        degree_labels = [
            r"\text{0}^{\circ}", 
            r"\text{90}^{\circ}", 
            r"\text{180}^{\circ}", 
            r"\text{270}^{\circ}", 
            r"\text{360}^{\circ}"
        ]
        dots = VGroup()
        tick_labels = VGroup()

        for theta, deg_label in zip(theta_values, degree_labels):
            phi = np.cos(theta)
            # Dot on the curve
            dot = Dot(point=axes.c2p(theta, phi), color=RED)
            dots.add(dot)

            # Label directly below the x-axis tick
            label = MathTex(deg_label).scale(0.5)
            label.next_to(axes.c2p(theta, 0), DOWN, buff=0.25)
            tick_labels.add(label)

        # Tracing dot
        tracing_dot = Dot(color=YELLOW).move_to(axes.c2p(0, np.cos(0)))
        

        # Create updater for the dot to follow the curve
        def update_dot(mob, alpha):
            x = interpolate(0, 2 * PI, alpha)
            y = np.cos(x)
            mob.move_to(axes.c2p(x, y))

        # Add the previous scene
        self.add(axes, axes_labels, dots, tick_labels, graph)

        # Show the tracing dot moving along the curve
        self.play(FadeIn(tracing_dot))
        self.play(UpdateFromAlphaFunc(tracing_dot, update_dot, run_time=5, rate_func=linear))
        self.wait()

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING flux_s3p4
```

# Transform Graph into the Equation

```python
class graphToEqn(Scene):
    def construct(self):
        # Set up axes
        axes = Axes(
            x_range=[0, 2 * PI, PI / 2],
            y_range=[-1.5, 1.5, 0.5],
            x_length=7,
            y_length=4,
            axis_config={"color": WHITE},
            tips=False
        )

        # Axis labels
        x_label = MathTex(r"\theta").next_to(axes.x_axis, RIGHT)
        y_label = MathTex(r"\Phi").next_to(axes.y_axis, UP)
        axes_labels = VGroup(x_label, y_label)

        # Plot the cosine curve: Phi(θ) = cos(θ)
        graph = axes.plot(lambda x: np.cos(x), color=BLUE)

        # Points and x-axis labels at key angles
        theta_values = [0, PI/2, PI, 3*PI/2, 2*PI]
        degree_labels = [
            r"\text{0}^{\circ}", 
            r"\text{90}^{\circ}", 
            r"\text{180}^{\circ}", 
            r"\text{270}^{\circ}", 
            r"\text{360}^{\circ}"
        ]
        dots = VGroup()
        tick_labels = VGroup()

        for theta, deg_label in zip(theta_values, degree_labels):
            phi = np.cos(theta)
            # Dot on the curve
            dot = Dot(point=axes.c2p(theta, phi), color=RED)
            dots.add(dot)

            # Label directly below the x-axis tick
            label = MathTex(deg_label).scale(0.5)
            label.next_to(axes.c2p(theta, 0), DOWN, buff=0.25)
            tick_labels.add(label)

        # Tracing dot
        tracing_dot = Dot(color=YELLOW).move_to(axes.c2p(0, np.cos(0)))
        

        # Create updater for the dot to follow the curve
        def update_dot(mob, alpha):
            x = interpolate(0, 2 * PI, alpha)
            y = np.cos(x)
            mob.move_to(axes.c2p(x, y))

        # Show everything
        self.play(Create(axes), Write(axes_labels))
        self.play(Create(graph))
        self.play(FadeIn(dots), FadeIn(tick_labels))
        self.play(UpdateFromAlphaFunc(tracing_dot, update_dot, run_time=5, rate_func=linear))
        self.wait(2)

        # Group all objects for transformation
        all_objects = VGroup(axes, axes_labels, graph, dots, tick_labels, tracing_dot)

        # Create the equation
        initial_equation = MathTex(r"\Phi \propto \cos(\theta)", font_size=72)

        # Transform the graph into the equation
        self.play(
            ReplacementTransform(all_objects, initial_equation),
            run_time=2
        )
        self.wait(2)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING graphToEqn
```

# Scene 4 Part 1
Creating the flux equation

```python
class flux_s4p1(Scene):
    def construct(self):
        # Step 1: Create the equation: Phi = cos(theta)
        initial_equation = MathTex(r"\Phi \propto \cos(\theta)", font_size=72)
        initial_equation.move_to(ORIGIN)
        self.play(Write(initial_equation))

        # Step 2: Then add E in front of cos
        updated_equationE = MathTex(r"\Phi \propto E \cos(\theta)", font_size=72)
        updated_equationE.move_to(initial_equation.get_center())

        self.play(ReplacementTransform(initial_equation, updated_equationE))
        self.wait()


# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING flux_s4p1
```

# Scene 4 Part 2
Turn into an equation

```python
class flux_s4p2(Scene):
    def construct(self):
        # Step 1: Create the equation: Phi = cos(theta)
        initial_equation = MathTex(r"\Phi \propto \cos(\theta)", font_size=72)
        initial_equation.move_to(ORIGIN)

        # Step 2: Then add E in front of cos
        updated_equationE = MathTex(r"\Phi \propto E \cos(\theta)", font_size=72)
        updated_equationE.move_to(initial_equation.get_center())

        # Add the updated equation directly
        self.add(updated_equationE)

        # Step 4: Turn into an equation
        updated_equationEA = MathTex(r"\Phi = EA \cos(\theta)", font_size=72)
        updated_equationEA.move_to(initial_equation.get_center())

        # Play the transformation
        self.play(ReplacementTransform(updated_equationE, updated_equationEA))
        self.wait()

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING flux_s4p2
```

# Scene 4 Part 3
The Flux Equation

```python
class flux_s4p3(Scene):
    def construct(self):
        # Step 2: Transform Phi into equation: Phi = cos(theta)
        initial_equation = MathTex(r"\Phi \propto \cos(\theta)", font_size=72)
        initial_equation.move_to(ORIGIN)  # Start from the same spot

        # Step 3: Then add E in front of cos
        updated_equationE = MathTex(r"\Phi \propto E \cos(\theta)", font_size=72)
        updated_equationE.move_to(initial_equation.get_center())

        # Step 4: Remove the alpha symbol and add in the equals sign and an A
        updated_equationEA = MathTex(r"\Phi = EA \cos(\theta)", font_size=72)
        updated_equationEA.move_to(initial_equation.get_center())

        # Step 5: Add the final equation with the correct notation
        # Final equation with E and A as vectors (bold)
        final_equation = MathTex(r"\Phi = \vec{E} \cdot \vec{A}", font_size=72)
        final_equation.move_to(initial_equation.get_center())

        self.play(Transform(updated_equationEA, final_equation))
        self.wait()

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING flux_s4p3
```
