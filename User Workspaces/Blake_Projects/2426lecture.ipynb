```python
from manim import *
import manim_physics as mp


class StaticFieldPic(Scene):
    def construct(self):
        # Create a grid of points
        number_plane = NumberPlane(
            background_line_style={
                "stroke_color": TEAL,
                "stroke_width": 2,
                "stroke_opacity": 0.6
            }
        )
        self.add(number_plane)

        x_min, x_max = -5, 5
        y_min, y_max = -3, 3
        n_charges = 10

        for _ in range(n_charges):
            x = np.random.uniform(x_min, x_max)
            y = np.random.uniform(y_min, y_max)
            magnitude = np.random.uniform(-3, 3)  # Random magnitude, positive or negative
            charge = mp.Charge(magnitude=magnitude, point=np.array([x, y, 0]), add_glow=False)
            field = mp.ElectricField(charge, color=YELLOW)
            self.add(charge, field)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING StaticFieldPic
```

```python
from manim import *
import manim_physics as mp


class LineOfCharge(Scene):
    def construct(self):
        # Create a grid of points
        number_plane = NumberPlane(
            background_line_style={
                "stroke_color": TEAL,
                "stroke_width": 2,
                "stroke_opacity": 0.6
            }
        )


        # Parameters
        x_min, x_max = -5, 5
        n_charges = 50  # Number of charges to add
        charges = []

        # Add charges at random x positions between x_min and x_max
        for _ in range(n_charges):
            x = np.random.uniform(x_min, x_max)
            charge = mp.Charge(magnitude=1, point=np.array([x, 0, 0]), add_glow = False)
            charges.append(charge)
            self.add(charge)

            # Remove previous field if it exists
            if len(charges) > 1:
                self.remove(field)

            # Show net field from all charges so far
            field = mp.ElectricField(*charges, color=YELLOW)
            self.add(field)
            self.wait(0.7)



# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING LineOfCharge
```
