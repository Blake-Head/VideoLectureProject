# A generic function for creating radial field lines

This loop can make any number of equally spaced lines that radiate outward from a central point

```python
class RadialLines(Scene):
    def construct(self):
        center_point = ORIGIN
        num_lines = 50
        radius = 3

        
        lines = VGroup()

        for i in range(num_lines):
            angle = i * (TAU / num_lines)
            end_point = radius * np.array([np.cos(angle), np.sin(angle), 0])
            line = Arrow(center_point, end_point)
            lines.add(line)

        self.play(FadeIn(lines))
        self.wait(2)
        self.play()
        self.play(FadeOut(lines))

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING RadialLines
```

# 3D Lines (Made with 2D Arrows)

This is a chearper version of the animation below which uses the Arrow3D object instead.

```python
class RadialLines3D(ThreeDScene):
    def construct(self):
        self.set_camera_orientation(phi=75 * DEGREES, theta=30 * DEGREES)

        center_point = ORIGIN
        num_lines = 6
        radius = 3

        lines = VGroup()
        for i in range(num_lines):
            for j in range(num_lines):
                theta = i * (TAU / num_lines)
                phi = j * (PI / num_lines)
                end_point = radius * np.array([
                    np.sin(phi) * np.cos(theta),
                    np.sin(phi) * np.sin(theta),
                    np.cos(phi)
                ])
                line = Arrow(center_point, end_point)

                lines.add(line)

        charge = Sphere(radius=2, 
                        color=RED, 
                        resolution=(48, 48), 
                        fill_opacity=0.5,
                        stroke_opacity=0.0,
                        )
        #self.add(lines)
        #self.play(Create(lines))
        self.play(FadeIn(charge))
        self.wait(2)
        self.play(GrowFromCenter(lines))
        self.wait(2)

        self.begin_ambient_camera_rotation(rate=0.5)
        self.wait(4)  # Adjust the duration to how long you want the rotation to last.
        self.stop_ambient_camera_rotation()

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING RadialLines3D
```

# Gaussian Blob

Here's a blob

```python
class GaussBlob(ThreeDScene):
    def construct(self):
        self.set_camera_orientation(phi=75 * DEGREES, theta=-45 * DEGREES)
        axes = ThreeDAxes()  # Add the XYZ coordinate axes

        # Add labels to the axes
        x_label = axes.get_x_axis_label("X")
        y_label = axes.get_y_axis_label("Y")
        z_label = axes.get_z_axis_label("Z")

        # Define the blob-like shape
        blob = Surface(
            lambda u, v: np.array([
                3 * np.sin(u) * np.cos(v) * (1 + 0.3 * np.sin(3 * u) * np.cos(3 * v)),
                2 * np.sin(u) * np.sin(v) * (1 + 0.3 * np.sin(3 * u) * np.cos(3 * v)),
                2 * np.cos(u) * (1 + 0.3 * np.sin(3 * u) * np.cos(3 * v))
            ]),
            u_range=[0, PI],
            v_range=[0, TAU],
            checkerboard_colors=[BLUE, GREEN]
        )

        self.add(axes, blob, x_label, y_label, z_label)
        
        self.begin_ambient_camera_rotation(rate=0.5)
        self.wait(10)  # Adjust the duration to how long you want the rotation to last.
        self.stop_ambient_camera_rotation()

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING GaussBlob
```
