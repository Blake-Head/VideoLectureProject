## (s1p1) Rotating Electric Field

```python
class s1p1(ThreeDScene):
    def construct(self):
        # Set camera orientation to view the plane above
        self.set_camera_orientation(phi=70 * DEGREES, theta=0 * DEGREES)

        # Place a box on top of the arrows
        box = Prism(dimensions=[15, 0.1, 15], fill_opacity=0.6, stroke_width=1).scale(0.5)
        box.set_fill(DARK_GREY, opacity=0.6)
        box.set_stroke(PURE_RED, width=1)
        box.move_to(ORIGIN)  # Lift it
        box.rotate(90*DEGREES, axis=RIGHT)

        grid = VGroup()
        rows, cols = 4, 4
        spacing = 1.5

        for i in range(rows):
            for j in range(cols):
                x = (j - (cols-1)/2) * spacing
                y = (i - (rows-1)/2) * spacing
                start = np.array([x, y, -2])
                end = np.array([x, y, 1])  # Arrow points up in z
                arrow = Arrow3D(start, end, color=BLUE, stroke_width=1)
                grid.add(arrow)


        # Begin ambient rotation to show off the scene
        self.add(grid)
        rotation_rate = 0.1
        rotation_angle = PI/2
        wait_time = rotation_angle / rotation_rate

        self.begin_ambient_camera_rotation(rate=rotation_rate)
        self.wait(wait_time)    
        #self.play(FadeIn(box), run_time=2)


# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING s1p1
```

## (s1p5) Fade to 2D Version

```python
class s1p5(ThreeDScene):
    def construct(self):
        # Set camera orientation to view the plane above
        self.set_camera_orientation(phi=90 * DEGREES, theta=0 * DEGREES)

        # Place a box on top of the arrows
        box = Prism(dimensions=[15, 0.1, 15], fill_opacity=0.6, stroke_width=1).scale(0.5)
        box.set_fill(DARK_GREY, opacity=0.6)
        box.set_stroke(PURE_RED, width=1)
        box.move_to(ORIGIN)  # Lift it
        box.rotate(90*DEGREES, axis=RIGHT)

        grid = VGroup()
        rows, cols = 4, 4
        spacing = 1.5

        for i in range(rows):
            for j in range(cols):
                x = (j - (cols-1)/2) * spacing
                y = (i - (rows-1)/2) * spacing
                start = np.array([x, y, -2])
                end = np.array([x, y, 1])  # Arrow points up in z
                arrow = Arrow3D(start, end, color=BLUE, stroke_width=1)
                grid.add(arrow)

        normal_arrow = Arrow3D(start=[0,0,0], end=[0,0,1], color=YELLOW, stroke_width=1)

        # Begin ambient rotation to show off the scene
        self.add(grid,box, normal_arrow)

        arrows = VGroup()  # Group to hold all arrows

        # Create 6 arrows spaced horizontally
        for i in range(6):
            arrow = Arrow(start=DOWN, end=UP*5, buff=0, color=BLUE)
            arrow.move_to(RIGHT * (i - 2.5) * 1.5)  # Center them nicely
            arrows.add(arrow)


        # Create the red line from the first to last arrow
        left_arrow = arrows[0]
        right_arrow = arrows[-1]
        red_line = Line(
            start=left_arrow.get_center(), 
            end=right_arrow.get_center(), 
            color=RED
        )

        # Place an arrow on the center of the red line
        center_arrow = Arrow(start=red_line.get_center(), end=UP*1.5, buff=0, color=YELLOW)
        # center_arrow = LabeledArrow("A", start=red_line.get_center(), end=UP, buff=0, color=YELLOW)

        # Add everything to the scene
        self.add_fixed_in_frame_mobjects(arrows, red_line, center_arrow)
        self.play(FadeIn(arrows), FadeIn(red_line), FadeIn(center_arrow),
                  FadeOut(grid, box, normal_arrow), run_time=2)
  
        self.wait()



# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING s1p5

```

## (s2p1) Rotate sheet

```python
class s2p1(Scene):
    def construct(self):
        arrows = VGroup()  # Group to hold all arrows

        # Create 6 arrows spaced horizontally
        for i in range(6):
            arrow = Arrow(start=DOWN, end=UP*5, buff=0, color=BLUE)
            arrow.move_to(RIGHT * (i - 2.5) * 1.5)  # Center them nicely
            arrows.add(arrow)


        # Create the red line from the first to last arrow
        left_arrow = arrows[0]
        right_arrow = arrows[-1]
        red_line = Line(
            start=left_arrow.get_center(), 
            end=right_arrow.get_center(), 
            color=RED
        )

        # Place an arrow on the center of the red line
        center_arrow = Arrow(start=red_line.get_center(), end=UP *1.5, buff=0, color=YELLOW)
        # center_arrow = LabeledArrow("A", start=red_line.get_center(), end=UP*1.5, buff=0, color=YELLOW)

        # Group the center arrow with the red line
        center_arrow_group = VGroup(red_line, center_arrow)

        # Add everything to the scene
        self.add(arrows)
        self.add(center_arrow_group)
        self.wait()


        # Create a dashed line attached to the center arrow
        dashed_line = DashedLine(
            start=center_arrow.get_start(), 
            end=center_arrow.get_start() + UP * 5, 
            color=WHITE, 
            dash_length=0.1
        )
         # Update the center arrow group to include the dashed line
        center_arrow_group.add(dashed_line)

        theta = -20 * DEGREES
        # Slightly rotate the center arrow group
        self.play(center_arrow_group.animate.rotate(theta, axis=OUT, about_point=red_line.get_center()))

        self.wait(2)

        # Calculate the coordinates for the dots along the rotated sheet
        coords = []
        for i in range(6):
            # Get the center of each arrow
            center = arrows[i].get_center()
            # Calculate the new y position using tan(theta)
            # Only rotate in the y direction, keep x and z the same
            rotated_center = center + np.array([0, np.tan(theta) * center[0], 0])
            coords.append(rotated_center)
        dots_and_lines = VGroup()
        for i in range(1, 5):
            dot = Dot(color=WHITE)
            dot.move_to(coords[i])
            line = Line(start=dot.get_center(), 
                        end=dot.get_center() + UP*1+RIGHT*1*np.sin(-theta), 
                        color=WHITE)
            self.play(FadeIn(dot, line), run_time=0.3)
            dots_and_lines.add(dot,line)
        self.wait(2)


        self.play(FadeOut(dots_and_lines))
        self.wait(1)

        

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING s2p1
```

## (s2p2) Rotate more

```python
class s2p2(Scene):
    def construct(self):
        arrows = VGroup()  # Group to hold all arrows

        # Create 6 arrows spaced horizontally
        for i in range(6):
            arrow = Arrow(start=DOWN, end=UP*5, buff=0, color=BLUE)
            arrow.move_to(RIGHT * (i - 2.5) * 1.5)  # Center them nicely
            arrows.add(arrow)


        # Create the red line from the first to last arrow
        left_arrow = arrows[0]
        right_arrow = arrows[-1]
        red_line = Line(
            start=left_arrow.get_center(), 
            end=right_arrow.get_center(), 
            color=RED
        )

        # Place an arrow on the center of the red line
        center_arrow = Arrow(start=red_line.get_center(), end=UP *1.5, buff=0, color=YELLOW)
        # center_arrow = LabeledArrow("A", start=red_line.get_center(), end=UP*1.5, buff=0, color=YELLOW)

        # Group the center arrow with the red line
        center_arrow_group = VGroup(red_line, center_arrow)

        # Create a dashed line attached to the center arrow
        dashed_line = DashedLine(
            start=center_arrow.get_start(), 
            end=center_arrow.get_start() + UP * 5, 
            color=WHITE, 
            dash_length=0.1
        )
        # Update the center arrow group to include the dashed line
        center_arrow_group.add(dashed_line)

        theta1 = -20 * DEGREES
        theta2 = -20 * DEGREES
        self.add(arrows)
        self.add(center_arrow_group.rotate(theta1, axis=OUT, about_point=red_line.get_center()))
        
        # Slightly rotate the center arrow group
        self.play(center_arrow_group.animate.rotate(theta2, axis=OUT, about_point=red_line.get_center()))

        self.wait(2)

        # Calculate the coordinates for the dots along the rotated sheet
        coords = []
        for i in range(6):
            # Get the center of each arrow
            center = arrows[i].get_center()
            # Calculate the new y position using tan(theta)
            # Only rotate in the y direction, keep x and z the same
            rotated_center = center + np.array([0, np.tan(theta1+theta2) * center[0], 0])
            coords.append(rotated_center)

        # Get the direction vector of the red line after rotation (using rotated endpoints)
        left_rot = coords[0]
        right_rot = coords[-1]
        red_line_dir = right_rot - left_rot
        red_line_dir = red_line_dir / np.linalg.norm(red_line_dir)

        # Perpendicular direction in the XY plane (rotate by 90 degrees)
        perp_dir = np.array([-red_line_dir[1], red_line_dir[0], 0])
        
        dots_and_lines = VGroup()

        for i in range(1, 5):
            dot = Dot(color=WHITE)
            dot.move_to(coords[i])
            line = Line(start=dot.get_center(), 
                        end=dot.get_center() + perp_dir, 
                        color=WHITE)
            self.play(FadeIn(dot, line), run_time=0.3)
            dots_and_lines.add(dot, line)
        self.wait(2)

        self.play(FadeOut(dots_and_lines))
        self.wait(1)




        

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING s2p2
```

```python
class s5p1(Scene):
    def construct(self):
        # Set up axes
        axes = Axes(
            x_range=[0, 2 * PI, PI / 2],
            y_range=[-1.5, 1.5, 0.5],
            x_length=7,
            y_length=4,
            axis_config={"color": WHITE},
            tips=False
        )

        # Axis labels
        x_label = MathTex(r"\theta").next_to(axes.x_axis, RIGHT)
        y_label = MathTex(r"\Phi").next_to(axes.y_axis, UP)
        axes_labels = VGroup(x_label, y_label)

        # Plot the cosine curve: Phi(θ) = cos(θ)
        graph = axes.plot(lambda x: np.cos(x), color=BLUE)

        # Points and x-axis labels at key angles
        theta_values = [PI/2, PI, 3*PI/2, 2*PI]
        degree_labels = [
            r"\text{90}^{\circ}", 
            r"\text{180}^{\circ}", 
            r"\text{270}^{\circ}", 
            r"\text{360}^{\circ}"
        ]
        dots = VGroup()
        tick_labels = VGroup()

        for theta, deg_label in zip(theta_values, degree_labels):
            phi = np.cos(theta)
            # Dot on the curve
            dot = Dot(point=axes.c2p(theta, phi), color=RED)
            dots.add(dot)

            # Label directly below the x-axis tick
            label = MathTex(deg_label).scale(0.5)
            label.next_to(axes.c2p(theta, 0), DOWN, buff=0.25)
            tick_labels.add(label)

        # Tracing dot
        tracing_dot = Dot(color=YELLOW).move_to(axes.c2p(0, np.cos(0)))
        

        # Create updater for the dot to follow the curve
        def update_dot(mob, alpha):
            x = interpolate(0, 2 * PI, alpha)
            y = np.cos(x)
            mob.move_to(axes.c2p(x, y))

        # Add the previous scene
        self.add(axes, axes_labels, dots, tick_labels, graph)
       # Create the equation
        initial_equation = MathTex(r"\Phi \propto \cos(\theta)", font_size=72).shift(UP * 2)
        # Show the tracing dot moving along the curve
        self.add(tracing_dot)
        self.play(UpdateFromAlphaFunc(tracing_dot, update_dot, run_time=5, rate_func=linear))
        self.play(UpdateFromAlphaFunc(tracing_dot, update_dot, run_time=5, rate_func=linear), Write(initial_equation))
        self.play(UpdateFromAlphaFunc(tracing_dot, update_dot, run_time=5, rate_func=linear))
        self.play(UpdateFromAlphaFunc(tracing_dot, update_dot, run_time=5, rate_func=linear))



        self.wait(2)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING s5p1
```

# Equation
Creating the flux equation

```python
class equation(Scene):
    def construct(self):
        # Step 1: Create the equation: Phi = cos(theta)
        initial_equation = MathTex(r"\Phi", r"\propto", r"\cos(\theta)", font_size=72)
        initial_equation.move_to(ORIGIN)
        self.play(Write(initial_equation))
        self.wait(1)

        # Step 2: Then add E in front of cos
        updated_equationE = MathTex(r"\Phi", r"\propto", r"E", r"\cos(\theta)", font_size=72)
        updated_equationE.move_to(initial_equation.get_center())

        # Step 3: Add in Area
        updated_equationEA = MathTex(r"\Phi", r"\propto", r"E", r"A", r"\cos(\theta)", font_size=72)
        updated_equationEA.move_to(initial_equation.get_center())

        # Step 4: Change propto to equal sign
        updated_equationEAEq = MathTex(r"\Phi", r"=", r"E", r"A", r"\cos(\theta)", font_size=72)
        updated_equationEA.move_to(initial_equation.get_center())

        self.play(TransformMatchingTex(initial_equation, updated_equationE))

        self.wait(2)
        self.play(TransformMatchingTex(updated_equationE, updated_equationEA))

        self.wait(2)
        self.play(TransformMatchingTex(updated_equationEA, updated_equationEAEq))

        self.wait(2)

        self.play(Indicate(updated_equationEAEq[2], scale_factor=1.8))

        self.wait(2)

        # Step 5: Change into a dot product
        final_equation = MathTex(r"\Phi", r"=", r"\vec{E}", r"\cdot", r"\vec{A}", font_size=72)
        final_equation.move_to(initial_equation.get_center())
        self.play(ReplacementTransform(updated_equationEAEq, final_equation))

        self.wait(2)


# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING equation
```

## (s6p1)

```python
class s6p1(ThreeDScene):
    def construct(self):
        self.set_camera_orientation(phi=70 * DEGREES, theta=30 * DEGREES)

        grid = VGroup()
        rows, cols = 3, 3
        spacing = 1

        for i in range(rows):
            for j in range(cols):
                x = (j - (cols-1)/2) * spacing
                y = (i - (rows-1)/2) * spacing
                start = np.array([x, y, -2])
                end = np.array([x, y, 1])  # Arrow points up in z
                arrow = Arrow3D(start, end, color=BLUE, stroke_width=1)
                grid.add(arrow)

        grid.move_to(UP * 3+LEFT * 1)
        self.add(grid)
        self.wait(1)


        # Place a box on top of the arrows
        box = Prism(dimensions=[4, 0.1, 4], fill_opacity=0.6, stroke_width=1)
        box.set_fill(DARK_GREY, opacity=0.6)
        box.set_stroke(PURE_RED, width=1)
        box.move_to(grid.get_center())  # Lift it
        box.rotate(90*DEGREES, axis=RIGHT)

        self.play(GrowFromCenter(box))
        self.wait(1)

        # Place a thin cylinder (circle)
        cylinder = Cylinder(radius=2, height=0.1, fill_opacity=0.6, stroke_width=1)
        cylinder.set_fill(DARK_GREY, opacity=0.6)
        cylinder.set_stroke(PURE_RED, width=1)
        cylinder.move_to(grid.get_center())  # Lift it
        cylinder.rotate(0*DEGREES, axis=RIGHT)

        self.play(ReplacementTransform(box, cylinder))
        #self.remove(box)
        #self.add(cylinder)
        self.wait(1)

        triangle1 = Polygon(
            np.array([-2, -2, 0]), 
            np.array([2, -2, 0]), 
            np.array([0, 2, 0]),
            fill_opacity=0.6, 
            stroke_width=1
        )
        triangle2 = Polygon(
                np.array([-2, -2, 0]), 
                np.array([2, -2, 0]), 
                np.array([0, 2, 0]),
                fill_opacity=0.6, 
                stroke_width=1
            )
        triangle1.set_fill(DARK_GREY, opacity=0.6)
        triangle1.set_stroke(PURE_RED, width=1)
        triangle1.move_to(grid.get_center()) 
        triangle1.rotate(0*DEGREES, axis=RIGHT) 

        triangle2.set_fill(DARK_GREY, opacity=0.6)
        triangle2.set_stroke(PURE_RED, width=1)
        triangle2.move_to(grid.get_center() + IN * 0.01)
        triangle2.rotate(0*DEGREES, axis=RIGHT) 

        triangle = VGroup(triangle1, triangle2)

        self.play(ReplacementTransform(cylinder, triangle))
        #self.remove(cylinder)
        #self.add(triangle)
        self.wait(2)

        self.play(FadeOut(triangle), FadeOut(grid))
        self.wait(1)






# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING s6p1
```
