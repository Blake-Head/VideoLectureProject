# Scene 1

## (s1p1) Plain Rotating Table

```python
class gauss_s1p1(ThreeDScene):
    def construct(self):
        #Set the camera position and orientation for a 3D view
        self.set_camera_orientation(phi=75 * DEGREES, theta=-45 * DEGREES)

        # Create a sheet (a thin rectangular prism) to represent the charged sheet
        sheet = Prism(dimensions=[4, 4, 0.1], fill_opacity=1, color=BLUE
        )

        # Create a simulated electric field around the sheet
        corners = [
            sheet.get_corner(UP + LEFT),
            sheet.get_corner(UP + RIGHT),
            sheet.get_corner(DOWN + LEFT),
            sheet.get_corner(DOWN + RIGHT)
        ]
        # Table legs as cylinders (or use Cuboid for rectangular legs)
        legs = VGroup()
        leg_height = 2
        leg_radius = 0.03
        for corner in corners:
            leg = Cylinder(
                radius=leg_radius,
                height=leg_height,
                direction=OUT,
                fill_opacity=1,
                color=GREY
            ).move_to(corner + IN* (leg_height / 2 + 0.05))  # 0.05 to account for tabletop thickness
            legs.add(leg)
        # Add the sheet, normal vector, and electric field arrows to the scene

        # Parameters for rotation and animation timing
        quarter_turn = PI/2
        rate = 0.25
        duration = quarter_turn / rate
        
        self.add(sheet, legs)
        self.begin_ambient_camera_rotation(rate=rate)
        self.wait(duration)  # Adjust the duration to how long you want the rotation to last.   
        self.stop_ambient_camera_rotation()

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING gauss_s1p1
    
```

## (s2p2) Spinning Parallel Normal Vector

```python
class gauss_s2p2(ThreeDScene):
    def construct(self):
        #Set the camera position and orientation for a 3D view
        self.set_camera_orientation(phi=0 * DEGREES, theta=0 * DEGREES)

        # Create a sheet (a thin rectangular prism) to represent the charged sheet
        sheet = Prism(dimensions=[4, 4, 0.1], fill_opacity=1, color=BLUE
        )

        # Create a normal vector to the sheet
        normal = always_redraw(lambda: Arrow3D(
            start=sheet.get_center(),
            end=sheet.get_center() + OUT * 1,
            color=YELLOW
        ))#.shift(UP * 1.5))  # Adjust the position slightly above the sheet

        pivot = sheet.get_center()
        arrow_length = 1.5

        angle_tracker = ValueTracker(180 * DEGREES)  # Start at 180 degrees

        def get_rotated_arrow():
            angle = angle_tracker.get_value()
            # Arrow starts at a point rotating around the pivot in the XY plane
            start = pivot + arrow_length * np.array([np.cos(angle), np.sin(angle), 0])
            return Arrow(
                start=sheet.get_center(),
                end=start,
                color=YELLOW,
                buff=-0.5
            )
        
        normal_para = always_redraw(get_rotated_arrow)

        # Create a simulated electric field around the sheet
        corners = [
            sheet.get_corner(UP + LEFT),
            sheet.get_corner(UP + RIGHT),
            sheet.get_corner(DOWN + LEFT),
            sheet.get_corner(DOWN + RIGHT)
        ]
        # Table legs as cylinders (or use Cuboid for rectangular legs)
        legs = VGroup()
        leg_height = 2
        leg_radius = 0.03
        for corner in corners:
            leg = Cylinder(
                radius=leg_radius,
                height=leg_height,
                direction=OUT,
                fill_opacity=1,
                color=GREY
            ).move_to(corner + IN* (leg_height / 2 + 0.05))  # 0.05 to account for tabletop thickness
            legs.add(leg)
        # Add the sheet, normal vector, and electric field arrows to the scene

        self.add(sheet, legs, normal, normal_para)
        self.play(angle_tracker.animate.set_value(3 * PI), run_time=4, rate_func=linear)




# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING gauss_s2p2
```

## (s3p2) Blake's Moving Vector

```python
class gauss_s3p2(ThreeDScene):
    def construct(self):
        self.set_camera_orientation(phi=80 * DEGREES, theta=45 * DEGREES)
        sheet = Prism(dimensions=[4, 4, 0.1], fill_opacity=1, color=BLUE)

        corners = [
            sheet.get_corner(UP + LEFT),
            sheet.get_corner(UP + RIGHT),
            sheet.get_corner(DOWN + LEFT),
            sheet.get_corner(DOWN + RIGHT)
        ]
        legs = VGroup()
        leg_height = 2
        leg_radius = 0.03
        for corner in corners:
            leg = Cylinder(
                radius=leg_radius,
                height=leg_height,
                direction=OUT,
                fill_opacity=1,
                color=GREY
            ).move_to(corner + IN * (leg_height / 2 + 0.05))
            legs.add(leg)

        # Get min/max x and y from corners
        xs = [corner[0] for corner in corners]
        ys = [corner[1] for corner in corners]
        min_x, max_x = min(xs), max(xs)
        min_y, max_y = min(ys), max(ys)

        # Trackers for position
        x_tracker = ValueTracker(sheet.get_center()[0])
        y_tracker = ValueTracker(sheet.get_center()[1])

        def get_normal():
            pos = np.array([x_tracker.get_value(), y_tracker.get_value(), 0])
            return Arrow3D(
                start=pos,
                end=pos + OUT * 1,
                color=YELLOW
            )

        def get_label():
            pos = np.array([x_tracker.get_value(), y_tracker.get_value(), 0])
            return MathTex(r"\hat{n}", color=YELLOW).scale(1.8).move_to(normal.get_end() + OUT * 0.9).rotate((1.5)*PI/2, axis=UP).rotate((1.00)*PI/2, axis=RIGHT).move_to(pos + OUT * 1.7)

        normal = always_redraw(get_normal)
        n_label = always_redraw(get_label)
        moving_vector = VGroup(normal, n_label)

        self.add(sheet, legs, moving_vector)

        # Animate the vector moving to each corner and center
        positions = [
            (min_x+3, min_y+2),
            (max_x-3, min_y+1),
            (max_x-1, max_y-2),
            (min_x+1, max_y-2),
            (sheet.get_center()[0], sheet.get_center()[1])
        ]
        for x, y in positions:
            self.play(
                x_tracker.animate.set_value(x),
                y_tracker.animate.set_value(y),
                run_time=1.2
            )
        self.wait(1)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING gauss_s3p2
```

## (s5p2) Multiple Arrows on Sphere

```python
class s5p1(ThreeDScene):
    def construct(self):
        self.set_camera_orientation(phi=80 * DEGREES, theta=45 * DEGREES)

        sphere = Sphere(radius=2)
        sphere.set_fill(BLUE, opacity=0.5)
        sphere.set_stroke(WHITE, width=0.75)

        self.add(sphere)
        #self.wait(2)

        # Place arrows at the center of each cell
        u_steps, v_steps = 16, 16  # controls density
        arrows = VGroup()
        for i in range(u_steps):
            u = (i + 0.5) / u_steps * PI  # latitude
            for j in range(v_steps):
                v = (j + 0.5) / v_steps * TAU  # longitude
                # Spherical to Cartesian
                x = sphere.radius * np.sin(u) * np.cos(v)
                y = sphere.radius * np.sin(u) * np.sin(v)
                z = sphere.radius * np.cos(u)
                pos = np.array([x, y, z])
                normal = pos / np.linalg.norm(pos)
                arrow = Arrow3D(
                    start=pos,
                    end=pos + normal * 0.5,
                    color=YELLOW,
                    stroke_width=1,
                )
                arrows.add(arrow)
        self.add(arrows)
        self.play(ShowIncreasingSubsets(arrows), run_time=3)
        self.wait(1)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING s5p1

import os
os.system('afplay /System/Library/Sounds/Glass.aiff')
```

```python
class s5p3(ThreeDScene):
    def construct(self):

        self.set_camera_orientation(phi=80 * DEGREES, theta=45 * DEGREES)

        sphere = Sphere(radius=2)
        sphere.set_fill(BLUE, opacity=0.5)
        sphere.set_stroke(WHITE, width=0.75)

        self.add(sphere)

        # Add two 2D normal vectors (arrows) starting on the surface of the sphere
        # Arrow 1: OUT direction (z+)
        normal1_start = sphere.get_center() + OUT * sphere.radius
        normal1_end = normal1_start + OUT * 2
        normal1 = Arrow(
            start=normal1_start,
            end=normal1_end,
            color=PURPLE_D,
            stroke_width=5,
        )

        # Arrow 2: LEFT direction (x-)
        normal2_start = sphere.get_center() - LEFT * sphere.radius
        normal2_end = normal2_start - LEFT * 2
        normal2 = Arrow(
            start=normal2_start,
            end=normal2_end,
            color=PURPLE,
            stroke_width=5,
        )

        # Dashed lines from sphere center to start of each arrow
        dashed1 = DashedLine(sphere.get_center(), normal1_start, color=PURPLE_D)
        dashed2 = DashedLine(sphere.get_center(), normal2_start, color=PURPLE)

        # Wait for a moment to show the sphere and normals
        # Create a 2D arrow that starts on the surface of the sphere
        arrow_start = sphere.get_center() + OUT * sphere.radius
        arrow_end = arrow_start + OUT * 2
        arrow = Arrow(
            start=arrow_start,
            end=arrow_end,
            color=YELLOW,
            stroke_width=5,
        )

        # Add the arrow
        #self.add(arrow)

        # Rotate the scene and move the arrow from equator to pole
        # Animate the arrow moving along the surface of the sphere from OUT to UP

        t = ValueTracker(0)

        def get_surface_point(t_val, s_val):
            # t_val: from OUT (z+) to UP (y+), phi from 0 to pi/2
            # s_val: from theta=0 to theta=pi/2 (move left along equator)
            phi = interpolate(0, PI / 2, t_val)
            theta = interpolate(0, PI / 2, s_val)
            x = np.sin(phi) * np.cos(theta)
            y = np.sin(phi) * np.sin(theta)
            z = np.cos(phi)
            return sphere.get_center() + np.array([x, y, z]) * sphere.radius

        t = ValueTracker(0)
        s = ValueTracker(0)

        def get_arrow():
            start = get_surface_point(t.get_value(), s.get_value())
            # Arrow points outward from the surface at this point
            direction = start - sphere.get_center()
            end = start + direction / np.linalg.norm(direction) * 2
            return Arrow3D(
            start=start,
            end=end,
            color=YELLOW,
            stroke_width=1,
            )

        moving_arrow = always_redraw(get_arrow)

        self.play(FadeIn(dashed1, dashed2))

        self.play(FadeIn(moving_arrow))
        self.wait(1)

        # Animate t (up) then s (left)
        self.play(t.animate.set_value(1), run_time=3, rate_func=linear)
        self.wait(1)
        self.play(s.animate.set_value(1), run_time=3, rate_func=linear)
        self.wait(2)

        self.play(FadeOut(moving_arrow, dashed1, dashed2), run_time=1)

# Load the manim extension
%load_ext manim

# Render the main scene in 4K quality
%manim -qk -v WARNING s5p3
```

```python
def make_grid_on_prism_top(center, x_len, y_len, n_lines=9, z_offset=0.05):
    grid = VGroup()
    for i in range(1, n_lines):
        frac = i / n_lines
        # Vertical lines (constant x, vary y)
        x = -x_len/2 + frac * x_len
        grid.add(Line(
            [x, -y_len/2, z_offset], [x, y_len/2, z_offset], color=WHITE, stroke_width=1
        ))
        # Horizontal lines (constant y, vary x)
        y = -y_len/2 + frac * y_len
        grid.add(Line(
            [-x_len/2, y, z_offset], [x_len/2, y, z_offset], color=WHITE, stroke_width=1
        ))
    grid.move_to(center + OUT * z_offset)
    return grid

class s6p1(ThreeDScene):
    def construct(self):

        self.set_camera_orientation(phi=75 * DEGREES, theta=45 * DEGREES)


        # Example: scale x and y independently
        x_len = 4
        y_len = 4
        sheet = Prism(dimensions=[x_len, y_len, 0.1], fill_opacity=1, color=BLUE)
        grid = make_grid_on_prism_top(sheet.get_center(), x_len, y_len, n_lines=9)
       
        # Create a normal vector to the sheet
        normal = always_redraw(lambda: Arrow3D(
            start=sheet.get_center(),
            end=sheet.get_center() + OUT * 1,
            color=YELLOW
        ))

        x_tracker = ValueTracker(sheet.get_center()[0])
        y_tracker = ValueTracker(sheet.get_center()[1])
        def get_label():
            pos = np.array([x_tracker.get_value(), y_tracker.get_value(), 0])
            return MathTex(r"\hat{n}", color=YELLOW).scale(1.8).move_to(normal.get_end() + OUT * 0.9).rotate((1.5)*PI/2, axis=UP).rotate((1.00)*PI/2, axis=RIGHT).move_to(pos + OUT * 1.7)
        
        n_label = always_redraw(get_label)

        self.add(sheet, grid, normal, n_label)
        
       # Trackers for scaling


        scale_x = ValueTracker(1)
        scale_y = ValueTracker(1)

        # Area label that scales with area
        def get_area_label():
            # Area is proportional to x_len * scale_x and y_len * scale_y
            area = x_len * scale_x.get_value() * y_len * scale_y.get_value()
           
            # Use sqrt(area) for scale so label grows with length, not area
            label_scale = 1.4 * np.sqrt(area) / 3 # adjust denominator for base size
            return MathTex(r"A", color=RED).scale(label_scale).next_to(n_label, 0.75*RIGHT + -0.75*UP, buff=0.05/area).rotate((1.5)*PI/2, axis=UP).rotate((1.00)*PI/2, axis=RIGHT)
        
        area_label = always_redraw(get_area_label)

        self.add(area_label)
        # Animate scaling

        self.play(
            scale_x.animate.set_value(3),
            scale_y.animate.set_value(1),
            sheet.animate.scale([3, 1, 1]),
            grid.animate.scale([3, 1, 1]),
            run_time=2
        )

        self.play(
            scale_x.animate.set_value(2),
            scale_y.animate.set_value(4),
            sheet.animate.scale([2/3, 4, 1]),  # relative to previous scale
            grid.animate.scale([2/3, 4, 1]), # relative to previous scale
            run_time=2
        )

        self.play(
            scale_x.animate.set_value(0.5),
            scale_y.animate.set_value(0.5),
            sheet.animate.scale([0.5/2, 0.5/4, 1]),  # relative to previous scale
            grid.animate.scale([0.5/2, 0.5/4, 1]), # relative to previous scale
            run_time=2
        )

        self.play(
            scale_x.animate.set_value(1),
            scale_y.animate.set_value(1),
            sheet.animate.scale([1/0.5, 1/0.5, 1]),  # relative to previous scale
            grid.animate.scale([1/0.5, 1/0.5, 1]), # relative to previous scale
            run_time=2
        )

        self.wait(1)



# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING s6p1
```

```python
class CylinderHW(ThreeDScene):
    def construct(self):
        # Set camera orientation for a better view of the cylinders
        self.set_camera_orientation(phi=75 * DEGREES, theta=5 * DEGREES)

        # Create a cylinder that is solid (normal)
        cylinder1 = Cylinder(radius=1.5, height=3, direction=OUT, show_ends=True)
        cylinder1.set_fill(BLUE, opacity=1)
        cylinder1.set_stroke(WHITE, width=0.5)
        cylinder1.shift(UP * 2)

        # Create a second cylinder that is hollow
        cylinder2 = Cylinder(radius=1.5, height=3, direction=OUT, show_ends=False)
        cylinder2.set_fill(RED, opacity=0.9)
        cylinder2.set_stroke(WHITE, width=0.5)

        cylinder2.shift(DOWN * 2)

        self.play(FadeIn(cylinder1, cylinder2))

       # Rotate the scene to show both cylinders clearly
        rate = 0.5
        rotation = 2 * PI
        duration = rotation / rate
        self.begin_ambient_camera_rotation(rate=rate)
        self.wait(duration)
        self.stop_ambient_camera_rotation()

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING CylinderHW
```

```python
class Logo(Scene):
    def construct(self):
        logo = SVGMobject("image2vector.svg").shift(1.25*RIGHT)
        logo2 = SVGMobject("file.svg").shift(1.25*LEFT)
        logos = VGroup(logo, logo2)

        font = 'Futura'
        color = YELLOW_E

        top_text = Text("Lion PER Physics Lectures", font_size=72, color=color, font=font).shift(UP * 2.5)
        bottom_text = Text("The Electric Field", font_size=72, color=color, font=font).shift(DOWN * 2.5)

        self.play(DrawBorderThenFill(logos), Write(top_text), run_time=1)
        self.wait(1)
        self.play(Write(bottom_text), run_time=1)
        self.wait(1)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk --transparent -v WARNING Logo

# Render the main scene
#%manim -qk -v WARNING Logo
```
