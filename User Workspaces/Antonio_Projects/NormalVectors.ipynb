# Plain Rotating Table (S1p1)

```python
class gauss_s1p1(ThreeDScene):
    def construct(self):
        #Set the camera position and orientation for a 3D view
        self.set_camera_orientation(phi=75 * DEGREES, theta=-45 * DEGREES)

        # Create a sheet (a thin rectangular prism) to represent the charged sheet
        sheet = Prism(dimensions=[4, 4, 0.1], fill_opacity=1, color=BLUE
        )

        # Create a simulated electric field around the sheet
        corners = [
            sheet.get_corner(UP + LEFT),
            sheet.get_corner(UP + RIGHT),
            sheet.get_corner(DOWN + LEFT),
            sheet.get_corner(DOWN + RIGHT)
        ]
        # Table legs as cylinders (or use Cuboid for rectangular legs)
        legs = VGroup()
        leg_height = 2
        leg_radius = 0.03
        for corner in corners:
            leg = Cylinder(
                radius=leg_radius,
                height=leg_height,
                direction=OUT,
                fill_opacity=1,
                color=GREY
            ).move_to(corner + IN* (leg_height / 2 + 0.05))  # 0.05 to account for tabletop thickness
            legs.add(leg)
        # Add the sheet, normal vector, and electric field arrows to the scene

        # Parameters for rotation and animation timing
        quarter_turn = PI/2
        rate = 0.25
        duration = quarter_turn / rate
        
        self.add(sheet, legs)
        self.begin_ambient_camera_rotation(rate=rate)
        self.wait(duration)  # Adjust the duration to how long you want the rotation to last.   
        self.stop_ambient_camera_rotation()

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING gauss_s1p1
    
```

# Spinning Parallel Normal Vector (S2p2)

```python
class gauss_s2p2(ThreeDScene):
    def construct(self):
        #Set the camera position and orientation for a 3D view
        self.set_camera_orientation(phi=0 * DEGREES, theta=0 * DEGREES)

        # Create a sheet (a thin rectangular prism) to represent the charged sheet
        sheet = Prism(dimensions=[4, 4, 0.1], fill_opacity=1, color=BLUE
        )

        # Create a normal vector to the sheet
        normal = always_redraw(lambda: Arrow3D(
            start=sheet.get_center(),
            end=sheet.get_center() + OUT * 1,
            color=YELLOW
        ))#.shift(UP * 1.5))  # Adjust the position slightly above the sheet

        pivot = sheet.get_center()
        arrow_length = 1.5

        angle_tracker = ValueTracker(180 * DEGREES)  # Start at 180 degrees

        def get_rotated_arrow():
            angle = angle_tracker.get_value()
            # Arrow starts at a point rotating around the pivot in the XY plane
            start = pivot + arrow_length * np.array([np.cos(angle), np.sin(angle), 0])
            return Arrow(
                start=sheet.get_center(),
                end=start,
                color=YELLOW,
                buff=-0.5
            )
        
        normal_para = always_redraw(get_rotated_arrow)

        # Create a simulated electric field around the sheet
        corners = [
            sheet.get_corner(UP + LEFT),
            sheet.get_corner(UP + RIGHT),
            sheet.get_corner(DOWN + LEFT),
            sheet.get_corner(DOWN + RIGHT)
        ]
        # Table legs as cylinders (or use Cuboid for rectangular legs)
        legs = VGroup()
        leg_height = 2
        leg_radius = 0.03
        for corner in corners:
            leg = Cylinder(
                radius=leg_radius,
                height=leg_height,
                direction=OUT,
                fill_opacity=1,
                color=GREY
            ).move_to(corner + IN* (leg_height / 2 + 0.05))  # 0.05 to account for tabletop thickness
            legs.add(leg)
        # Add the sheet, normal vector, and electric field arrows to the scene

        self.add(sheet, legs, normal, normal_para)
        self.play(angle_tracker.animate.set_value(3 * PI), run_time=4, rate_func=linear)




# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING gauss_s2p2
```

# Blake's Moving Vector (S3p2)

```python
class gauss_s3p2(ThreeDScene):
    def construct(self):
        self.set_camera_orientation(phi=80 * DEGREES, theta=45 * DEGREES)
        sheet = Prism(dimensions=[4, 4, 0.1], fill_opacity=1, color=BLUE)

        corners = [
            sheet.get_corner(UP + LEFT),
            sheet.get_corner(UP + RIGHT),
            sheet.get_corner(DOWN + LEFT),
            sheet.get_corner(DOWN + RIGHT)
        ]
        legs = VGroup()
        leg_height = 2
        leg_radius = 0.03
        for corner in corners:
            leg = Cylinder(
                radius=leg_radius,
                height=leg_height,
                direction=OUT,
                fill_opacity=1,
                color=GREY
            ).move_to(corner + IN * (leg_height / 2 + 0.05))
            legs.add(leg)

        # Get min/max x and y from corners
        xs = [corner[0] for corner in corners]
        ys = [corner[1] for corner in corners]
        min_x, max_x = min(xs), max(xs)
        min_y, max_y = min(ys), max(ys)

        # Trackers for position
        x_tracker = ValueTracker(sheet.get_center()[0])
        y_tracker = ValueTracker(sheet.get_center()[1])

        def get_normal():
            pos = np.array([x_tracker.get_value(), y_tracker.get_value(), 0])
            return Arrow3D(
                start=pos,
                end=pos + OUT * 1,
                color=YELLOW
            )

        def get_label():
            pos = np.array([x_tracker.get_value(), y_tracker.get_value(), 0])
            return MathTex(r"\hat{n}", color=YELLOW).scale(1.8).move_to(normal.get_end() + OUT * 0.9).rotate((1.5)*PI/2, axis=UP).rotate((1.00)*PI/2, axis=RIGHT).move_to(pos + OUT * 1.7)

        normal = always_redraw(get_normal)
        n_label = always_redraw(get_label)
        moving_vector = VGroup(normal, n_label)

        self.add(sheet, legs, moving_vector)

        # Animate the vector moving to each corner and center
        positions = [
            (min_x+3, min_y+2),
            (max_x-3, min_y+1),
            (max_x-1, max_y-2),
            (min_x+1, max_y-2),
            (sheet.get_center()[0], sheet.get_center()[1])
        ]
        for x, y in positions:
            self.play(
                x_tracker.animate.set_value(x),
                y_tracker.animate.set_value(y),
                run_time=1.2
            )
        self.wait(1)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING gauss_s3p2
```

# Sphere with moving Normal Vector part 1

```python
class SphereNormal1(ThreeDScene):
    def construct(self):
        # Create the sphere
        sphere = Sphere(radius=2)
        sphere.set_fill(BLUE, opacity=0.5)
        sphere.set_stroke(WHITE, width=0.75)

        # Set camera orientation
        self.set_camera_orientation(phi=60 * DEGREES, theta=45 * DEGREES)
        self.add(sphere)

        # Add two 2D normal vectors (arrows) starting on the surface of the sphere
        # Arrow 1: OUT direction (z+)
        normal1_start = sphere.get_center() + OUT * sphere.radius
        normal1_end = normal1_start + OUT * 2
        normal1 = LabeledArrow(
            start=normal1_start,
            end=normal1_end,
            color=PURPLE_D,
            stroke_width=5,
            label=MathTex(r"\hat{r}"),
            label_position=0.6
        )

        # Arrow 2: LEFT direction (x-)
        normal2_start = sphere.get_center() - LEFT * sphere.radius
        normal2_end = normal2_start - LEFT * 2
        normal2 = LabeledArrow(
            start=normal2_start,
            end=normal2_end,
            color=PURPLE,
            stroke_width=5,
            label=MathTex(r"\hat{r}"),
            label_position=0.6
        )

        # Dashed lines from sphere center to start of each arrow
        dashed1 = DashedLine(sphere.get_center(), normal1_start, color=PURPLE_D)
        dashed2 = DashedLine(sphere.get_center(), normal2_start, color=PURPLE)
        self.add(dashed1, dashed2, 
                 normal1, normal2
        )

        # Wait for a moment to show the sphere and normals
        self.wait(3)

        # Create a 2D arrow that starts on the surface of the sphere
        arrow_start = sphere.get_center() + OUT * sphere.radius
        arrow_end = arrow_start + OUT * 2
        arrow = LabeledArrow(
            start=arrow_start,
            end=arrow_end,
            color=YELLOW,
            stroke_width=5,
            label=MathTex(r"\hat{r}"),
            label_position=0.6
        )

        # Add the arrow
        self.add(arrow)

        # Rotate the scene and move the arrow from equator to pole
        self.begin_ambient_camera_rotation(rate=0.5)

        # Animate the arrow moving along the surface of the sphere from OUT to UP

        t = ValueTracker(0)

        def get_surface_point(t_val, s_val):
            # t_val: from OUT (z+) to UP (y+), phi from 0 to pi/2
            # s_val: from theta=0 to theta=pi/2 (move left along equator)
            phi = interpolate(0, PI / 2, t_val)
            theta = interpolate(0, PI / 2, s_val)
            x = np.sin(phi) * np.cos(theta)
            y = np.sin(phi) * np.sin(theta)
            z = np.cos(phi)
            return sphere.get_center() + np.array([x, y, z]) * sphere.radius

        t = ValueTracker(0)
        s = ValueTracker(0)

        def get_arrow():
            start = get_surface_point(t.get_value(), s.get_value())
            # Arrow points outward from the surface at this point
            direction = start - sphere.get_center()
            end = start + direction / np.linalg.norm(direction) * 2
            return LabeledArrow(
            start=start,
            end=end,
            color=YELLOW,
            stroke_width=5,
            label=MathTex(r"\hat{r}"),
            label_position=0.6
            )

        moving_arrow = always_redraw(get_arrow)
        self.add(moving_arrow)

        # Remove the initial static arrow before animating the yellow arrow
        self.remove(arrow)

        # Animate t (up) then s (left)
        self.play(t.animate.set_value(1), run_time=3, rate_func=linear)
        self.play(s.animate.set_value(1), run_time=3, rate_func=linear)
        self.wait(2)

# Load the manim extension
%load_ext manim

# Render the main scene in 4K quality
%manim -qm -v WARNING SphereNormal1
```

# Rotating Cube Scene
Scene of a rotating cube with 2D normal vectors fading in

```python
class CubeNormal(ThreeDScene):
    def construct(self):
        # Set camera orientation to show 3 faces clearly
        self.set_camera_orientation(phi=90 * DEGREES, theta=45 * DEGREES)

        # Create a cube with larger faces
        cube = Cube(side_length=2.5)
        #cube.set_fill(BLUE, opacity=1.0)
        cube.set_stroke(WHITE, width=1)

        # Add the cube to the scene

        # Rotate the scene to show different faces
        self.begin_ambient_camera_rotation(rate=0.5)
        self.wait(1)  # Adjust the duration to how long you want the rotation to last.

        # Add labeled 2D arrows to each face (using Arrow instead of Arrow3D)
        arrows = VGroup()
        face_normals = [OUT, IN, RIGHT, LEFT, UP, DOWN]
        face_colors = [YELLOW, PURE_RED, GREEN, PURPLE, BLUE, WHITE]
        face_rotations = [0, np.pi, -np.pi/2, np.pi/2, np.pi/2, -np.pi/2]
        arrow_list = []
        for normal, color, rot in zip(face_normals, face_colors, face_rotations):
            face_center = cube.get_center() + normal * (cube.side_length / 2)
            start_3d = face_center
            end_3d = face_center + normal * 1.5
            arrow = Arrow3D(
            start=start_3d,
            end=end_3d,
            color=color,
            )
            if rot != 1:
                arrow.rotate(rot, axis=normal, about_point=face_center)
                arrow_list.append(arrow)
                arrows.add(arrow)

        # Parameters for rotation and animation timing
        turn = 2*PI
        rate = 0.25
        duration = turn / rate
        fade_time = 0.5

        self.add(cube)
        # Fade in arrows one by one
        for arrow in arrow_list:
            self.play(FadeIn(arrow, run_time=fade_time))

        # Wait to show the arrows
        # Rotate the scene again to show the arrows
        self.begin_ambient_camera_rotation(rate=rate)
        self.wait(duration)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING CubeNormal
```

```python
class CubeNormal(ThreeDScene):
    def construct(self):
        self.set_camera_orientation(phi=90 * DEGREES, theta=45 * DEGREES)

        cube = Cube(side_length=2.5)
        cube.set_fill(BLUE, opacity=0.5)
        cube.set_stroke(WHITE, width=1)
        self.add(cube)

        self.begin_ambient_camera_rotation(rate=0.5)
        self.wait(1)

        # Normals and setup
        face_normals = [OUT, IN, RIGHT, LEFT, UP, DOWN]
        face_colors = [YELLOW, RED, GREEN, PURPLE, BLUE, WHITE]
        face_rotations = [0, np.pi, -np.pi/2, np.pi/2, np.pi/2, -np.pi/2]

        arrows = VGroup()
        labels = VGroup()

        for normal, color, rot in zip(face_normals, face_colors, face_rotations):
            face_center = cube.get_center() + normal * (cube.side_length / 2)
            arrow = Arrow(
                start=face_center,
                end=face_center + normal * 1.5,
                color=color,
                stroke_width=6
            )
            arrow.rotate(rot, axis=normal, about_point=face_center)
            arrows.add(arrow)

            # Create an always-facing label in matching color
            def make_label(arrow_ref=arrow, norm_dir=normal, arrow_color=color):
                return always_redraw(
                    lambda: MathTex(r"\hat{n}", color=arrow_color)
                    .move_to(arrow_ref.get_end() + norm_dir * 0.25)
                    .rotate(self.camera.get_theta(), axis=LEFT)
                    # self.add.fix.in frame()
                        # use this to fix the n hat to the camera's frame
                    
                )
            
            label = make_label()
            labels.add(label)

        # Animate arrows and labels
        for arrow, label in zip(arrows, labels):
            self.play(FadeIn(arrow), FadeIn(label), run_time=0.5)

        self.wait(2)
        self.begin_ambient_camera_rotation(rate=0.5)
        self.wait(10)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING CubeNormal
```

# Two Cylinder HW Problem Scene

```python
class CylinderHW(ThreeDScene):
    def construct(self):
        # Set camera orientation for a better view of the cylinders
        self.set_camera_orientation(phi=75 * DEGREES, theta=5 * DEGREES)

        # Create a cylinder that is solid (normal)
        cylinder1 = Cylinder(radius=1.5, height=3, direction=OUT, show_ends=True)
        cylinder1.set_fill(BLUE, opacity=1)
        cylinder1.set_stroke(WHITE, width=0.5)
        cylinder1.shift(UP * 2)

        # Create a second cylinder that is hollow
        cylinder2 = Cylinder(radius=1.5, height=3, direction=OUT, show_ends=False)
        cylinder2.set_fill(RED, opacity=0.9)
        cylinder2.set_stroke(WHITE, width=0.5)

        cylinder2.shift(DOWN * 2)

        self.add(cylinder1, cylinder2)

        self.wait(2)

        # Rotate the scene to show both cylinders clearly
        self.begin_ambient_camera_rotation(rate=0.5)

        self.wait(5)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qm -v WARNING CylinderHW
```

# Create a growing/shrinking Sheet

```python
class DynamicSheet(ThreeDScene):
    def construct(self):
        # Set the camera position and orientation for a 3D view
        self.set_camera_orientation(phi=75 * DEGREES, theta=-45 * DEGREES)

        # Create a sheet (a thin rectangular prism) to represent the charged sheet
        sheet = Prism(dimensions=[4, 4, 0.1], fill_opacity=1, color=BLUE)
        self.add(sheet)

        # Attach a normal vector n to the sheet
        normal = Arrow(
            start=sheet.get_center(),
            end=sheet.get_center() + OUT * 2,
            color=YELLOW,
            buff=0
        )
        self.add(normal)

        # Add the area label "A" to the sheet
        area_label = MathTex("A", color=RED).scale(1.5)
        area_label.move_to(sheet.get_center() + UP * 1.5 + RIGHT * 1.5)

        # Wait a few seconds before changing the arrow
        self.wait(5)

        # Create a new normal to simulate changing color
        normalBlue = Arrow(
            start=sheet.get_center(),
            end=sheet.get_center() + OUT * 2,
            color=DARK_BLUE,
            buff=0
        )
        self.remove(normal)
        self.add(normalBlue)


        # Create a VGroup for the normal and sheet
        sheet_group = VGroup(sheet, normalBlue)
        self.add(sheet_group)

        # Grow the sheet and Arrow dynamically
        self.play(sheet_group.animate.scale(1.5), run_time=5, rate_func=there_and_back)
        self.wait(2)

        # Shrink the sheet dynamically
        self.play(sheet_group.animate.scale(0.5), run_time=5, rate_func=there_and_back)
        self.wait(2)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING DynamicSheet
```
