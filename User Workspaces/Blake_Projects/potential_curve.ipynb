```python
class Energy(Scene):
    def construct(self):
        # Create axes for potential energy graph
          # Create axes for potential energy graph
        axes_potential = Axes(
            x_range=[-5, 5, 1],  # X-axis range
            y_range=[0, 5, 1],   # Y-axis range
            axis_config={"color": WHITE},  # Axis color
        ).scale(0.5).shift(3.5 * LEFT + 2 * UP)  # Scale down and shift to the top left

        # Add labels to the potential energy graph
        labels_potential = axes_potential.get_axis_labels(x_label="x(m)", y_label="U(J)").scale(0.8)
        k = 1
        # Define the potential energy parabola
        parabola_potential = axes_potential.plot(lambda x: 0.5 * k * x**2, color=BLUE, x_range=[-3, 3])

        # Add a title for the potential energy graph
        title_potential = MathTex("U_s = \\frac{1}{2}kx^2", color=WHITE).next_to(axes_potential, DOWN).scale(0.8)

        # Create a dot for the potential energy graph
        dot_potential = Dot(color=GREEN, radius=0.1).move_to(axes_potential.input_to_graph_point(2.5, parabola_potential))

        # Create axes for kinetic energy graph
        axes_kinetic = Axes(
            x_range=[-5, 5, 1],  # X-axis range
            y_range=[0, 5, 1],   # Y-axis range
            axis_config={"color": WHITE},  # Axis color
        ).scale(0.5).shift(3.5 * RIGHT + 2 * UP)  # Scale down and shift to the top right

        # Add labels to the kinetic energy graph
        labels_kinetic = axes_kinetic.get_axis_labels(x_label="v(m/s)", y_label="K(J)").scale(0.8)
        m = 1
        # Define the kinetic energy parabola
        parabola_kinetic = axes_kinetic.plot(lambda v: 0.5 * m * v**2, color=GREEN, x_range=[-3, 3])

        # Add a title for the kinetic energy graph
        title_kinetic = MathTex("K = \\frac{1}{2}mv^2", color=WHITE).next_to(axes_kinetic, DOWN).scale(0.8)

        # Create a dot for the kinetic energy graph
        dot_kinetic = Dot(color=GREEN, radius=0.1).move_to(axes_kinetic.input_to_graph_point(0, parabola_kinetic))

        # Add everything to the scene
        self.play(Create(axes_potential), Write(labels_potential))
        self.play(Create(parabola_potential), Write(title_potential))
        self.play(FadeIn(dot_potential))

        self.play(Create(axes_kinetic), Write(labels_kinetic))
        self.play(Create(parabola_kinetic), Write(title_kinetic))
        self.play(FadeIn(dot_kinetic))

        self.wait(1)

        self.play(Flash(dot_potential), Flash(dot_kinetic))
        self.wait(1)

# Add a curly spring and a mass in the bottom half of the scene
        spring = ParametricFunction(
            lambda t: np.array([
                t,  # x-coordinate
                0.2 * np.sin(10 * t),  # y-coordinate (sinusoidal wave)
                0  # z-coordinate
            ]),
            t_range=[0, 2],  # Length of the spring
            color=WHITE
        ).shift(2 * DOWN)  # Position the spring

        mass = Square(side_length=0.5, color=RED, fill_opacity=1).move_to(spring, RIGHT)

        # Add the spring and mass to the scene
        self.play(Create(spring), FadeIn(mass))

        self.wait(2)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING Energy
```

```python
class Energy2(Scene):
    def construct(self):
        # Create axes for potential energy graph
        axes_potential = Axes(
            x_range=[-5, 5, 1],  # X-axis range
            y_range=[0, 5, 1],   # Y-axis range
            axis_config={"color": WHITE},  # Axis color
        ).scale(0.5).shift(3.5 * LEFT + 2 * UP)  # Scale down and shift to the top left

        # Add labels to the potential energy graph
        labels_potential = axes_potential.get_axis_labels(x_label="x(m)", y_label="U(J)").scale(0.8)
        k = 1
        # Define the potential energy parabola
        parabola_potential = axes_potential.plot(lambda x: 0.5 * k * x**2, color=BLUE, x_range=[-3, 3])

        # Add a title for the potential energy graph
        title_potential = MathTex("U_s = \\frac{1}{2}kx^2", color=WHITE).next_to(axes_potential, DOWN).scale(0.8)

        # Create a dot for the potential energy graph
        dot_potential = Dot(color=GREEN, radius=0.1)

        # Create axes for kinetic energy graph
        axes_kinetic = Axes(
            x_range=[-5, 5, 1],  # X-axis range
            y_range=[0, 5, 1],   # Y-axis range
            axis_config={"color": WHITE},  # Axis color
        ).scale(0.5).shift(3.5 * RIGHT + 2 * UP)  # Scale down and shift to the top right

        # Add labels to the kinetic energy graph
        labels_kinetic = axes_kinetic.get_axis_labels(x_label="v(m/s)", y_label="K(J)").scale(0.8)
        m = 1
        # Define the kinetic energy parabola
        parabola_kinetic = axes_kinetic.plot(lambda v: 0.5 * m * v**2, color=GREEN, x_range=[-3, 3])

        # Add a title for the kinetic energy graph
        title_kinetic = MathTex("K = \\frac{1}{2}mv^2", color=WHITE).next_to(axes_kinetic, DOWN).scale(0.8)

        # Create a dot for the kinetic energy graph
        dot_kinetic = Dot(color=GREEN, radius=0.1)

        # Add everything to the scene
        self.play(Create(axes_potential), Write(labels_potential))
        self.play(Create(parabola_potential), Write(title_potential))
        self.play(FadeIn(dot_potential))

        self.play(Create(axes_kinetic), Write(labels_kinetic))
        self.play(Create(parabola_kinetic), Write(title_kinetic))
        self.play(FadeIn(dot_kinetic))

        self.wait(1)

        # Add a curly spring and a mass in the bottom half of the scene
        spring = ParametricFunction(
            lambda t: np.array([
                t,  # x-coordinate
                0.2 * np.sin(10 * t),  # y-coordinate (sinusoidal wave)
                0  # z-coordinate
            ]),
            t_range=[0, 2],  # Length of the spring
            color=WHITE
        ).shift(2 * DOWN)

        mass = Square(side_length=0.5, color=RED, fill_opacity=1).next_to(spring, RIGHT)

        # Add the spring and mass to the scene
        self.play(Create(spring), FadeIn(mass))

        total_energy = 5  # Total energy of the system

        # Oscillation animation
        def update_mass_and_spring(mass, dt):
            t = self.renderer.time % 4  # Oscillation period
            x = 2 * np.cos(PI * t)  # Oscillation equation
            mass.move_to(np.array([x, -2, 0]))  # Update mass position

        def update_dots(dot):
            t = self.renderer.time % 4
            x = 2 * np.cos(PI * t)
            v = -2 * PI * np.sin(PI * t)
            potential_energy = 0.5 * k * x**2
            kinetic_energy = total_energy - potential_energy  # Enforce constant total energy
            if dot == dot_potential:
                dot.move_to(axes_potential.c2p(x, potential_energy))
            elif dot == dot_kinetic:
                dot.move_to(axes_kinetic.c2p(v, kinetic_energy))

        mass.add_updater(update_mass_and_spring)
        dot_potential.add_updater(update_dots)
        dot_kinetic.add_updater(update_dots)

        self.wait(10)  # Let the animation run for 10 seconds

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING Energy2
```
