# Moving Blocks

```python

class s1p1(Scene):
    def construct(self):

        number_plane = NumberPlane(
            background_line_style={
                "stroke_color": BLUE_D,
                "stroke_width": 4,
                "stroke_opacity": 0.1,

            }
        )# Axis labels
        x_label = MathTex("x").next_to(number_plane.c2p(6, 0), RIGHT*1.5+DOWN*0.7)
        y_label = MathTex("y").next_to(number_plane.c2p(0, 3), UP+RIGHT*0.7+UP*0.2)

        self.add(number_plane, x_label, y_label)

        # ── Change these freely ──────────────────────────────────────────────
        g_mass = 0.5      # green block mass  (also sets visual side_length)
        r_mass = 2    # red block mass
        vg_i   = 2.0    # initial speed of green (manim-units per anim-second)
        vr_i   = 0    # initial speed of red   (0 = stationary)

        g_start = 6.0   # how far left  of origin the green center starts
        r_start = 2.0   # how far right of origin the red   center starts
        T_post  = 3.0   # how many anim-seconds to show post-collision motion
        # ────────────────────────────────────────────────────────────────────

        # 1D elastic collision equations
        vg_f = ((g_mass - r_mass) * vg_i + 2 * r_mass * vr_i) / (g_mass + r_mass)
        vr_f = ((r_mass - g_mass) * vr_i + 2 * g_mass * vg_i) / (g_mass + r_mass)

        # Ground line sits at y = -2; each block is raised so its bottom edge
        # touches the ground regardless of its size.
        ground_y = 2
        small_green = Square(side_length=g_mass, color=GREEN, fill_opacity=0.5).shift(
            LEFT * g_start + DOWN * (ground_y - g_mass / 2)
        )
        green_label = MathTex("m_g").next_to(small_green, DOWN)
        green_velocity_arrow_pre = Arrow(
            small_green.get_center(),
            small_green.get_center() + RIGHT * vg_i,
            buff=0,
            color=GREEN
        )
        g_vel_label_pre = MathTex("v_{gi}").next_to(green_velocity_arrow_pre, UP)
        big_red = Square(side_length=r_mass, color=RED, fill_opacity=0.5).shift(
            RIGHT * r_start + DOWN * (ground_y - r_mass / 2)
        )
        red_label = MathTex("m_r").next_to(big_red, DOWN)
        ground_line = Line(LEFT * 8, RIGHT * 8).shift(DOWN * ground_y)

        self.add(ground_line, small_green, big_red, green_label, red_label, g_vel_label_pre, green_velocity_arrow_pre)

        # Pre-collision: both blocks move at their initial velocities until they meet.
        # The gap between their facing edges closes at the relative velocity (vg_i - vr_i).
        gap = (g_start - g_mass / 2) + (r_start - r_mass / 2)
        t_collision   = gap / (vg_i - vr_i)
        dist_green_pre = vg_i * t_collision
        dist_red_pre   = vr_i * t_collision

        self.play(
            small_green.animate.shift(RIGHT * dist_green_pre),
            green_label.animate.shift(RIGHT * dist_green_pre),
            green_velocity_arrow_pre.animate.shift(RIGHT * dist_green_pre),
            big_red.animate.shift(RIGHT * dist_red_pre),
            red_label.animate.shift(RIGHT * dist_red_pre),
            g_vel_label_pre.animate.shift(RIGHT * dist_green_pre),
            run_time=t_collision,
            rate_func=linear
        )

        # Post-collision: both blocks move simultaneously at their final velocities
        # for T_post seconds.  distance = |v_final| * T_post
        dist_green_post = abs(vg_f) * T_post
        dist_red_post   = abs(vr_f) * T_post

        dir_green = LEFT if vg_f < 0 else RIGHT
        dir_red   = RIGHT if vr_f > 0 else LEFT

        green_velocity_arrow_post = Arrow(
            small_green.get_center(),
            small_green.get_center() + dir_green * abs(vg_f),
            buff=0,
            color=GREEN
        )
        g_vel_label_post = MathTex("v_{gf}").next_to(green_velocity_arrow_post, UP)
        self.add(green_velocity_arrow_post, g_vel_label_post)
        self.remove(green_velocity_arrow_pre, g_vel_label_pre)

        red_velocity_arrow_post = Arrow(
            big_red.get_center(),
            big_red.get_center() + dir_red * abs(vr_f),
            buff=0,
            color=RED
        )
        r_vel_label_post = MathTex("v_{rf}").next_to(red_velocity_arrow_post, UP)
        self.add(red_velocity_arrow_post, r_vel_label_post)

        self.play(
            small_green.animate.shift(dir_green * dist_green_post),
            green_label.animate.shift(dir_green * dist_green_post),
            green_velocity_arrow_post.animate.shift(dir_green * dist_green_post),
            g_vel_label_post.animate.shift(dir_green * dist_green_post),
            big_red.animate.shift(dir_red   * dist_red_post),
            red_label.animate.shift(dir_red   * dist_red_post),
            red_velocity_arrow_post.animate.shift(dir_red * dist_red_post),
            r_vel_label_post.animate.shift(dir_red * dist_red_post),
            run_time=T_post,
            rate_func=linear
        )

        self.wait(1)


# Load the manim extension
%load_ext manim

# Render the main scene
%manim -qk -v WARNING s1p1
```
