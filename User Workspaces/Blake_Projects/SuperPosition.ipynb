# Basic Superposition Things
In this block, we make a few charges and their relevant electric fields.

Each def lays out part of the overall scene, with everything being played piece by piece at the end in the def construct(self): piece.

I think this is a better way of putting the entire animation together since we can run individual pieces without needing to render the entire file.


```python
class SuperPosition(Scene):
    def create_charges(self):
        self.charge_1 = mp.Charge(magnitude=1, add_glow=0).move_to(RIGHT*2)
        self.charge_2 = mp.Charge(magnitude=-1, add_glow=0).move_to(LEFT*2)
        self.play(FadeIn(self.charge_1), FadeIn(self.charge_2))

    def create_fields(self):
        self.field_1 = always_redraw(lambda: mp.ElectricField(self.charge_1))
        self.field_2 = always_redraw(lambda: mp.ElectricField(self.charge_2))
        self.play(FadeIn(self.field_1), FadeIn(self.field_2))

    def combine_fields(self):
        self.field_3 = always_redraw(lambda: mp.ElectricField(self.charge_1, self.charge_2))
        self.play(FadeIn(self.field_3))

    def construct(self):
        self.create_charges()
        self.create_fields()
        self.combine_fields()

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING SuperPosition
```

# ZoomedScenes and Camera Controls
Here, I'm testing how to use a ZoomedScene for when we want to show how the vectors add up.

For now, I'm putting an xy axis in the frame so I know what coordinates to move the camera to.

```python
# In order to use zooming in features, must used ZoomedScene class
class ZoomTest(ZoomedScene):
    def construct(self):
        charge_1 = mp.Charge(magnitude=1, add_glow=0).move_to(RIGHT*2)
        charge_2 = mp.Charge(magnitude=-1, add_glow=0).move_to(LEFT*2)
        field_1 = always_redraw(lambda: mp.ElectricField(charge_1, color=RED))
        field_2 = always_redraw(lambda: mp.ElectricField(charge_2, color=BLUE))
        field_3 = always_redraw(lambda: mp.ElectricField(charge_1, charge_2))

        plane = (
            NumberPlane(x_range=[-15,15,1], y_range=[-15,15,1])
            .add_coordinates()
        )

        self.add(charge_1, charge_2, field_1, field_2, plane)

        self.activate_zooming(animate=True)
        self.play(self.zoomed_camera.frame.animate.shift(2.45*UP, 2.55*LEFT), function=smooth, run_time = 2)
        self.play(self.zoomed_camera.frame.animate.shift(1.95*DOWN, 5.05*RIGHT), function=smooth, run_time = 2)
        self.wait(2)
        self.play(FadeIn(field_3))
        self.wait(2)
        self.play(FadeOut(field_1, field_2))
        self.wait(1)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING ZoomTest
```

# Vector Addition
Let's make sure we know how to show vector addition!

Ideally, we should be able to show for any of the zoomed in vector pairs above how they add to make the resultant vector. 

```python
class VectorAddition(Scene):
    def construct(self):
        plane = (
        NumberPlane(x_range=[-15,15,1], y_range=[-15,15,1])
        .add_coordinates()
        )   
        v1 = Vector(2*RIGHT, color=RED)
        v2 = Vector(2*UP, color=BLUE)
        self.add(plane)
        self.play(FadeIn(v1), FadeIn(v2))

        self.wait(2)
        #self.play(v2.animate.shift(2*RIGHT))
        v3 = always_redraw(lambda: Vector(v1.get_end() + v2.get_end(), color=GREEN))
        self.play(FadeIn(v3))

        self.play(v1.animate.scale(2, about_point=ORIGIN))
        self.wait(2)
        self.play(v2.animate.scale(-1, about_point=ORIGIN))
        self.wait(2)



# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING VectorAddition
```

# Multiple Charges and Complex Fields

At some point we'll want to show how adding more and more charges leads to more and more complicated fields.

Co-pilot made some loops for me to make the actual code more efficient and compact, but you could totally write this brute-force style, likely doesn't really matter for the kinds of stuff we are doing.

```python
class ManyCharges(Scene):
    def construct(self):
        # Generate random positions for the charges
        positions = [np.array([random.uniform(-4, 4), random.uniform(-4, 4), 0]) for _ in range(10)]

        # Create charges at random positions
        charges = [mp.Charge(magnitude=1 if i % 2 == 0 else -1, add_glow=0).move_to(pos) for i, pos in enumerate(positions)]

        # Create electric fields for the charges
        fields = [mp.ElectricField(*charges[:i+1]) for i in range(len(charges))]

        self.play(FadeIn(charges[0], fields[0]))
        for i in range(9):
            self.play(FadeIn(charges[i+1]), Transform(fields[0], fields[i+1]))
            self.wait(1)

        self.wait(2)
# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING ManyCharges
```

```python
class RadialLines(Scene):
    def construct(self):
        center_point = ORIGIN
        num_lines = 24
        radius = 3

        lines = VGroup()
        for i in range(num_lines):
            angle = i * (TAU / num_lines)
            end_point = radius * np.array([np.cos(angle), np.sin(angle), 0])
            line = Line(center_point, end_point)
            lines.add(line)

        self.play(Create(lines))
        self.wait(2)

# Load the manim extension
%load_ext manim

# Render the main scene
%manim -ql -v WARNING RadialLines
```
